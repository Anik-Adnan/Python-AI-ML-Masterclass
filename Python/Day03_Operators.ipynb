## 🔷 Section 1 — Comparison Operators

Comparison operators evaluate a relationship between two values and return a **`bool`**.

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `<` | Less than | `3 < 5` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<=` | Less than or equal | `5 <= 5` | `True` |
| `>=` | Greater than or equal | `6 >= 5` | `True` |

### Important: `==` vs `=`
- `=` is the **assignment operator**: `x = 5` binds name x to object 5
- `==` is the **equality operator**: `x == 5` tests whether x equals 5

---

````python
# ============================================================
# COMPARISON OPERATORS
# ============================================================

a, b = 10, 20
print(f"a={a}, b={b}")
print(f"a == b  : {a == b}")    # False
print(f"a != b  : {a != b}")    # True
print(f"a <  b  : {a <  b}")    # True
print(f"a >  b  : {a >  b}")    # False
print(f"a <= b  : {a <= b}")    # True
print(f"a >= b  : {a >= b}")    # False

# ── String comparison uses Unicode code point ordering ──
print("\nString comparison (lexicographic):")
print(f"'apple' < 'banana' : {'apple' < 'banana'}")   # True  (a < b)
print(f"'Zebra' < 'apple'  : {'Zebra' < 'apple'}")    # True  (Z=90 < a=97 in Unicode)
print(f"'abc' == 'abc'     : {'abc' == 'abc'}")        # True

# ── Comparing different types ──
print("\nCross-type comparisons:")
print(f"1 == True  : {1 == True}")     # True  (True == 1 because bool is int)
print(f"0 == False : {0 == False}")    # True
print(f"1 == 1.0   : {1 == 1.0}")     # True  (int vs float: values compared)
# print(1 < "2")   # TypeError in Python 3 — can't compare int and str

# ── Comparison result is always bool ──
result = 10 > 5
print(f"\ntype(10 > 5) = {type(result)}")   # <class 'bool'>
````

## 🔷 Section 2 — Logical Operators

| Operator | Meaning | Returns |
|----------|---------|---------|
| `and` | True only if **both** operands are truthy | First falsy value, or last value |
| `or` | True if **at least one** operand is truthy | First truthy value, or last value |
| `not` | Negation | `True` or `False` (always a bool) |

### Short-Circuit Evaluation

Python's `and` and `or` do **not** necessarily evaluate both operands.
They stop as soon as the result is determined — this is called **short-circuiting**.

```
and  →  returns FIRST FALSY value,  or last value if all truthy
or   →  returns FIRST TRUTHY value, or last value if all falsy
```

This has important consequences:
1. **Performance**: avoid expensive computations in the second operand when not needed
2. **Safety**: `if x is not None and x > 0:` won't evaluate `x > 0` if `x is None`
3. **Idioms**: `value = user_input or "default"` — provide a default value

---

````python
# ============================================================
# LOGICAL OPERATORS — and, or, not
# ============================================================

# ── Truth tables ──
print("AND truth table:")
for a in [True, False]:
    for b in [True, False]:
        print(f"  {str(a):<5} and {str(b):<5} = {a and b}")

print("\nOR truth table:")
for a in [True, False]:
    for b in [True, False]:
        print(f"  {str(a):<5} or  {str(b):<5} = {a or b}")

print("\nNOT:")
print(f"  not True  = {not True}")
print(f"  not False = {not False}")
print(f"  not 0     = {not 0}")
print(f"  not []    = {not []}")
````

````python
# ============================================================
# SHORT-CIRCUIT EVALUATION — CRITICAL CONCEPT
# ============================================================

# ── 'and' returns the FIRST FALSY value (or last value if all truthy) ──
print("and short-circuit:")
print(0 and print("never printed"))   # 0 — stops at 0, never evaluates right
print(1 and "hello")                  # "hello" — both truthy, returns last
print(False and 1/0)                  # False — right side never evaluated (no ZeroDivisionError!)

print()

# ── 'or' returns the FIRST TRUTHY value (or last value if all falsy) ──
print("or short-circuit:")
print(0 or "default")    # "default" — 0 is falsy, returns next
print("found" or "alt")  # "found"   — truthy, returns it immediately
print(0 or False or None or "last")  # "last" — all previous were falsy

print()

# ── Practical idioms using short-circuit ──
# Pattern 1: default value
username = ""
display = username or "Guest"
print(f"Display name: {display}")    # Guest (empty string is falsy)

# Pattern 2: guard clause (avoid None dereference)
data = None
length = data is not None and len(data)  # Won't call len(None)
print(f"Safe length: {length}")          # False (short-circuited)

# Pattern 3: conditional execution
debug_mode = True
debug_mode and print("DEBUG: checkpoint reached")   # Only prints if debug_mode is truthy
````

## 🔷 Section 3 — Bitwise Operators

Bitwise operators work on the **binary representations** of integers, one bit at a time.
They are used in systems programming, cryptography, embedded systems, and performance-critical code.

| Operator | Name | Example | Result | Description |
|----------|------|---------|--------|-------------|
| `&` | AND | `12 & 10` | `8` | 1 only where BOTH bits are 1 |
| `\|` | OR | `12 \| 10` | `14` | 1 where EITHER bit is 1 |
| `^` | XOR | `12 ^ 10` | `6` | 1 where bits DIFFER |
| `~` | NOT | `~12` | `-13` | Flip all bits (`~n = -(n+1)`) |
| `<<` | Left shift | `1 << 4` | `16` | Multiply by 2 per shift |
| `>>` | Right shift | `16 >> 2` | `4` | Integer divide by 2 per shift |

### Binary Representation

```
12 = 0b1100   (8 + 4)
10 = 0b1010   (8 + 2)
   &
    = 0b1000  = 8   (AND: only bit 3 is set in both)
```

---

````python
# ============================================================
# BITWISE OPERATORS
# ============================================================

a = 12   # binary: 1100
b = 10   # binary: 1010

print(f"a = {a:4d}  ({a:08b})")
print(f"b = {b:4d}  ({b:08b})")
print(f"{'─'*25}")
print(f"a & b  = {a&b:3d}  ({a&b:08b})  AND")
print(f"a | b  = {a|b:3d}  ({a|b:08b})  OR")
print(f"a ^ b  = {a^b:3d}  ({a^b:08b})  XOR")
print(f"~a     = {~a:3d}  ({~a & 0xFF:08b})  NOT (8-bit view)")
print(f"a << 2 = {a<<2:3d}  ({a<<2:08b})  Left shift 2")
print(f"a >> 1 = {a>>1:3d}  ({a>>1:08b})  Right shift 1")

print()
# ── Practical applications ──
# Check if a number is even using bit 0
n = 47
print(f"{n} is {'even' if (n & 1) == 0 else 'odd'}  (last bit = {n & 1})")

# Fast power-of-2 check: n & (n-1) == 0 for powers of 2
for x in [1, 2, 3, 4, 7, 8, 16, 31, 32]:
    is_pow2 = x > 0 and (x & (x-1)) == 0
    print(f"  {x:3d} is{'    ' if is_pow2 else ' NOT'} a power of 2")

# Left/right shift as multiply/divide by 2
print(f"\n7 << 3 = {7 << 3} (= 7 × 8)")
print(f"100 >> 2 = {100 >> 2} (= 100 ÷ 4)")
````

## 🔷 Section 4 — Identity and Membership Operators

### Identity Operators: `is` and `is not`

`is` tests whether two names refer to the **exact same object** in memory.
`==` tests whether two objects have **equal values**.

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

a == b    # True  — same values
a is b    # False — different objects (different id())
a is c    # True  — c is the same object as a (same id())
```

#### When to use `is`
- Comparing to `None`: always use `x is None` (not `x == None`)
- Comparing to singletons: `True`, `False`, `None`
- Never use `is` to compare numbers or strings in general code

### Membership Operators: `in` and `not in`

`in` tests whether a value exists within a sequence or collection.
Works with `str`, `list`, `tuple`, `set`, `dict` (tests keys), and any iterable.

---

````python
# ============================================================
# IDENTITY OPERATORS — is, is not
# ============================================================

# ── is vs == ──
a = [1, 2, 3]
b = [1, 2, 3]   # Same values, different object
c = a           # Same object

print(f"a == b  : {a == b}")    # True  (equal values)
print(f"a is b  : {a is b}")    # False (different objects, different id)
print(f"a is c  : {a is c}")    # True  (same object, same id)
print(f"id(a)={id(a)}, id(b)={id(b)}, id(c)={id(c)}")

print()

# ── None checking — always use 'is None' ──
x = None
print(f"x is None     : {x is None}")      # True  — correct idiom
print(f"x == None     : {x == None}")      # True  — works but not idiomatic
print(f"x is not None : {x is not None}")  # False

print()

# ── Small integer caching (CPython implementation detail) ──
# CPython caches integers from -5 to 256. Outside this range, new objects
# are created even for equal values.
p, q = 42, 42
r, s = 1000, 1000
print(f"42 is 42       : {p is q}")    # True  (cached)
print(f"1000 is 1000   : {r is s}")    # False in standard Python (not cached)
# Note: Jupyter/Colab may show True due to compile-time optimisation.
# The key lesson: NEVER rely on 'is' for value comparison of integers or strings.
````

````python
# ============================================================
# MEMBERSHIP OPERATORS — in, not in
# ============================================================

# ── in with strings ──
sentence = "The quick brown fox jumps"
print(f"'fox' in sentence      : {'fox' in sentence}")       # True
print(f"'cat' in sentence      : {'cat' in sentence}")       # False
print(f"'The' in sentence      : {'The' in sentence}")       # True (case-sensitive)
print(f"'the' in sentence      : {'the' in sentence}")       # False

print()

# ── in with lists ──
primes = [2, 3, 5, 7, 11, 13, 17, 19]
print(f"7 in primes   : {7 in primes}")    # True
print(f"9 in primes   : {9 in primes}")    # False
print(f"9 not in primes: {9 not in primes}") # True

print()

# ── in with dict (tests KEYS, not values) ──
student = {"name": "Alice", "age": 20, "gpa": 3.8}
print(f"'name' in student  : {'name' in student}")    # True  (key exists)
print(f"'Alice' in student : {'Alice' in student}")   # False (value, not key)
print(f"20 in student      : {20 in student}")        # False

print()

# ── in with a set (O(1) average lookup — much faster than list for large data) ──
valid_colors = {"red", "green", "blue", "yellow"}
user_choice = "purple"
print(f"'{user_choice}' is valid: {user_choice in valid_colors}")
````

## 🔷 Section 5 — Chained Comparisons & The Walrus Operator

### Chained Comparisons

Python allows comparison operators to be **chained** in a mathematically natural way:

```python
# Instead of:
if x > 0 and x < 10:
    ...

# Write:
if 0 < x < 10:    # Reads like mathematics — much more Pythonic
    ...
```

The chained expression `a < b < c` is equivalent to `a < b and b < c`,
but `b` is only evaluated **once**.

### The Walrus Operator `:=` (Python 3.8+, PEP 572)

The walrus operator assigns a value to a variable **as part of an expression**:

```python
# Without walrus: read twice, compute once
data = get_data()
if data:
    process(data)

# With walrus: assign and test in one step
if data := get_data():
    process(data)
```

Use walrus sparingly — only where it genuinely improves readability.

---

````python
# ============================================================
# CHAINED COMPARISONS
# ============================================================

# ── Mathematical-style range checks ──
x = 15
print(f"x = {x}")
print(f"0 < x < 100          : {0 < x < 100}")           # True
print(f"10 <= x <= 20        : {10 <= x <= 20}")          # True
print(f"x == 15 == 15        : {x == 15 == 15}")          # True
print(f"0 < x < 10           : {0 < x < 10}")             # False

# ── Chaining with different operators ──
a, b, c = 1, 2, 3
print(f"\na < b < c           : {a < b < c}")             # True
print(f"a < b > c            : {a < b > c}")              # False (b > c? 2 > 3? No)
print(f"a != b != c          : {a != b != c}")            # True (a≠b AND b≠c)

# ── Grade classification using chained comparisons ──
def letter_grade(score):
    if score < 0 or score > 100:
        return "Invalid"
    elif 90 <= score <= 100: return "A"
    elif 80 <= score < 90:  return "B"
    elif 70 <= score < 80:  return "C"
    elif 60 <= score < 70:  return "D"
    else:                    return "F"

for s in [95, 83, 70, 65, 45, 101]:
    print(f"  Score {s:3d} → {letter_grade(s)}")
````

````python
# ============================================================
# THE WALRUS OPERATOR := (Python 3.8+)
# ============================================================
# The walrus operator (:=) assigns a value to a name
# as a side-effect inside an expression.
# Name comes from its visual resemblance to a walrus face.

import re

# ── Walrus in a while loop (classic use case) ──
# Traditional approach:
numbers = [4, 7, 2, 9, 1, 8, 3]
idx = 0
while idx < len(numbers):
    val = numbers[idx]
    if val > 5:
        print(f"Found > 5 at index {idx}: {val}")
    idx += 1

print()

# ── Walrus to avoid double computation ──
data = [1, -2, 3, -4, 5, 6, -7, 8]

# Find and process values > 4 in one expression
large = [y for x in data if (y := x * 2) > 4]   # (Day 24: list comprehension)
print(f"Doubled values > 4: {large}")

# ── Walrus in a conditional expression ──
import math
values = [16, -4, 25, 9, -1]
for v in values:
    if v >= 0 and (sq := math.sqrt(v)) == int(sq):   # guard avoids ValueError on negatives
        print(f"sqrt({v}) = {sq:.1f} — perfect square")
    elif v < 0:
        print(f"sqrt({v}) is undefined for real numbers (negative input)")
````

## 🔷 Section 6 — Complete Operator Precedence Table

From highest (evaluated first) to lowest (evaluated last):

| Priority | Operators | Description |
|----------|-----------|-------------|
| 1 (highest) | `()` | Parentheses — override everything |
| 2 | `**` | Exponentiation (right-associative) |
| 3 | `+x`, `-x`, `~x` | Unary plus, minus, bitwise NOT |
| 4 | `*`, `/`, `//`, `%` | Multiplicative |
| 5 | `+`, `-` | Additive |
| 6 | `<<`, `>>` | Bitwise shifts |
| 7 | `&` | Bitwise AND |
| 8 | `^` | Bitwise XOR |
| 9 | `\|` | Bitwise OR |
| 10 | `==`, `!=`, `<`, `>`, `<=`, `>=`, `is`, `is not`, `in`, `not in` | Comparisons |
| 11 | `not` | Logical NOT |
| 12 | `and` | Logical AND |
| 13 (lowest) | `or` | Logical OR |

**Rule of thumb**: when in doubt, add parentheses. They never hurt readability.

---

````python
# ============================================================
# OPERATOR PRECEDENCE — Worked Examples
# ============================================================

# ── Common precedence pitfalls ──
print("Precedence examples:")
print(f"2 + 3 * 4        = {2 + 3 * 4}")          # 14 (not 20)
print(f"not 5 > 3        = {not 5 > 3}")           # False (> before not)
print(f"not (5 > 3)      = {not (5 > 3)}")         # False (same, explicit)
print(f"True or False and False = {True or False and False}")  # True (and before or)
print(f"(True or False) and False = {(True or False) and False}")  # False

print()
# ── Precedence with bitwise ──
print(f"2 + 3 << 1       = {2 + 3 << 1}")    # 10: (2+3)=5 then 5<<1=10
print(f"2 + (3 << 1)     = {2 + (3 << 1)}")  # 8:  3<<1=6 then 2+6=8
# Note: + has higher priority than <<? Let's check:
# Actually << has lower priority than +, so 2 + 3 << 1 = (2+3) << 1 = 10

print()
# ── Building complex boolean expressions ──
age, income, credit_score = 25, 55000, 720

# Loan approval logic: (age >= 18 and age <= 65) and income > 40000 and credit_score >= 700
eligible = (18 <= age <= 65) and income > 40_000 and credit_score >= 700
print(f"Loan eligible: {eligible}")

# Equivalent with explicit operator priority
eligible2 = ((18 <= age <= 65) and (income > 40_000) and (credit_score >= 700))
print(f"Same result  : {eligible2}")
````

````python
# EXERCISE 3 — Bitwise Operations
# Implement a simple permission system using bitbit flags (Unix-style)

READ    = 0b100   # 4
WRITE   = 0b010   # 2
EXECUTE = 0b001   # 1

# Grant permissions using OR
user_perms  = READ | WRITE      # 6 = 0b110
admin_perms = READ | WRITE | EXECUTE  # 7 = 0b111
guest_perms = READ              # 4 = 0b100

def has_perm(perms, flag):
    return bool(perms & flag)

for name, perms in [("User", user_perms), ("Admin", admin_perms), ("Guest", guest_perms)]:
    r = has_perm(perms, READ)
    w = has_perm(perms, WRITE)
    x = has_perm(perms, EXECUTE)
    print(f"{name:<6}: read={r}, write={w}, execute={x}  (binary: {perms:03b})")

# Revoke WRITE from admin
admin_no_write = admin_perms & ~WRITE
print(f"\nAdmin without write: {admin_no_write:03b} = {admin_no_write}")
````

````python
# EXERCISE 4 — Membership Testing
# Build a simple spell-checker using 'in' with a set

dictionary = {"apple", "banana", "cherry", "date", "elderberry",
              "fig", "grape", "honeydew", "kiwi", "lemon"}

words_to_check = ["apple", "Banana", "cherry", "GRAPE", "kiwi", "mango"]

print("Spell Check Results:")
print(f"{'Word':<12} {'Lowercase':<12} {'Found'}")
print("-" * 35)
for word in words_to_check:
    lower = word.lower()
    found = lower in dictionary
    status = "✓" if found else "✗ (not found)"
    print(f"{word:<12} {lower:<12} {status}")

# Count how many are valid
valid = sum(1 for w in words_to_check if w.lower() in dictionary)
print(f"\n{valid}/{len(words_to_check)} words found in dictionary")
````

````python
# EXERCISE 5 — Operator Precedence Prediction
# Before running, predict each result. Then verify.

expressions = [
    ("2 ** 3 ** 2",           2 ** 3 ** 2),       # Right-associative: 2**(3**2)=512
    ("not 3 > 2",             not 3 > 2),          # > before not: not True = False
    ("1 + 2 == 3 and 4 > 3",  1 + 2 == 3 and 4 > 3),  # +,==,> before and → True
    ("False or not False",    False or not False), # not first: not False=True → True
    ("4 & 6 | 2",             4 & 6 | 2),          # &=4, 4|2=6
    ("~3 + 4",                ~3 + 4),             # ~3=-4, -4+4=0
    ("3 < 5 != 4 < 6",        3 < 5 != 4 < 6),    # chained: True and True = True
    ("True + True * 2",       True + True * 2),   # *=2, True+2=3
]

print(f"{'Expression':<30} {'Result'}")
print("-" * 42)
for expr, result in expressions:
    print(f"  {expr:<28} = {result}")
````

````python
# CHALLENGE 1 — Compound Interest with Walrus Operator
# A = P(1 + r/n)^(nt)
# Find the first year when investment doubles.

principal = 10_000.0
annual_rate = 0.07    # 7%
n = 12               # Compounded monthly

year = 0
balance = principal

print(f"Principal: ${principal:,.2f} at {annual_rate*100}% APR, compounded monthly")
print(f"{'Year':<6} {'Balance':>12} {'Doubled?'}")
print("-" * 32)
while year <= 30:
    balance = principal * (1 + annual_rate/n) ** (n * year)
    doubled = balance >= 2 * principal
    if year % 5 == 0 or doubled:
        print(f"{year:<6} ${balance:>11,.2f} {'✓ DOUBLED!' if doubled else ''}")
    if doubled:
        break
    year += 1

print(f"\nRule of 72 estimate: ~{72 / (annual_rate*100):.1f} years")
````

````python
# CHALLENGE 2 — Simple XOR Cipher (Bitwise Encryption)
# XOR cipher: encrypt by XORing each character with a key byte.
# Decryption is identical — XOR is its own inverse.

def xor_cipher(text, key):
    result = []
    for char in text:
        result.append(chr(ord(char) ^ key))
    return "".join(result)

key = 42   # Secret key (0-127 recommended for printable ASCII)
plaintext  = "Hello, Python!"

encrypted  = xor_cipher(plaintext, key)
decrypted  = xor_cipher(encrypted, key)   # XOR twice = original

print(f"Key       : {key} ({bin(key)})")
print(f"Plaintext : {plaintext!r}")
print(f"Encrypted : {encrypted!r}")
print(f"Decrypted : {decrypted!r}")
print(f"Correct   : {decrypted == plaintext}")

# ── Show byte-level XOR for first 3 chars ──
print("\nByte-level detail (first 3 chars):")
for c in plaintext[:3]:
    enc = ord(c) ^ key
    print(f"  '{c}' ({ord(c):3d} = {ord(c):08b})  XOR {key:08b}  = {enc:08b} = {enc:3d} '{chr(enc)}'")
````
