## Section 1 - Adding Elements

| Method | Description | Returns |
|--------|-------------|---------|
| append(x) | Add single element x to the end | None |
| extend(iterable) | Add all elements of iterable to the end | None |
| insert(i, x) | Insert x before index i | None |

### append vs extend - Common Pitfall

```python
lst = [1, 2, 3]
lst.append([4, 5])   # [1, 2, 3, [4, 5]]  - ONE element (a list)
lst.extend([4, 5])   # [1, 2, 3, 4, 5]    - TWO elements added
```
---

````python
# ============================================================
# ADDING ELEMENTS - append, extend, insert
# ============================================================

lst = [1, 2, 3]
lst.append(4)
print(f"After append(4)   : {lst}")
lst.append([5, 6])
print(f"After append([5,6]): {lst}")

lst = [1, 2, 3]
lst.extend([4, 5, 6])
print(f"\nAfter extend([4,5,6]): {lst}")
lst.extend("abc")
print(f"After extend('abc') : {lst}")
lst.extend(range(10, 13))
print(f"After extend(range) : {lst}")

fruits = ["apple", "cherry", "date"]
fruits.insert(1, "banana")
print(f"\nAfter insert(1,banana): {fruits}")
fruits.insert(0, "avocado")
print(f"After insert(0,avocado): {fruits}")
fruits.insert(len(fruits), "elderberry")
print(f"After insert at end: {fruits}")

result = lst.append(99)
print(f"\nappend() returns: {result}")
````

## Section 2 - Removing Elements

| Method | Description | Returns |
|--------|-------------|---------|
| remove(x) | Remove first occurrence of x; ValueError if missing | None |
| pop(i=-1) | Remove and return element at index i; default last | The removed element |
| clear() | Remove all elements | None |
| del lst[i] | Delete element at index (statement, not a method) | - |
---

````python
# ============================================================
# REMOVING ELEMENTS - remove, pop, clear, del
# ============================================================

lst = [10, 20, 30, 20, 40, 20]
print(f"Original: {lst}")

lst.remove(20)
print(f"After remove(20)  : {lst}")
lst.remove(20)
print(f"After remove(20)  : {lst}")

try:
    lst.remove(99)
except ValueError as e:
    print(f"remove(99) error  : {e}")

lst = [10, 20, 30, 40, 50]
last   = lst.pop()
print(f"\npopped last : {last},    list={lst}")
first  = lst.pop(0)
print(f"popped [0]  : {first},   list={lst}")
middle = lst.pop(1)
print(f"popped [1]  : {middle},   list={lst}")

temp = [1, 2, 3, 4, 5]
temp.clear()
print(f"\nAfter clear(): {temp}")

lst = [10, 20, 30, 40, 50, 60]
del lst[2]
print(f"\nAfter del [2]: {lst}")
del lst[1:3]
print(f"After del [1:3]: {lst}")
````

## Section 4 - Sorting and Reversing

| Operation | In-place? | Returns |
|-----------|-----------|---------|
| lst.sort() | Yes - modifies lst | None |
| sorted(lst) | No - original unchanged | New sorted list |
| lst.reverse() | Yes - modifies lst | None |
| reversed(lst) | No - original unchanged | Iterator |

### The key Parameter

sort() and sorted() accept a key function applied to each element before
comparison.

```python
students.sort(key=lambda s: s["gpa"], reverse=True)
```
---

````python
# ============================================================
# SORTING AND REVERSING - sort vs sorted
# ============================================================

nums = [5, 2, 8, 1, 9, 3, 7, 4, 6]

copy1 = nums.copy()
copy1.sort()
print(f"sort()   (ascending) : {copy1}")

copy2 = nums.copy()
copy2.sort(reverse=True)
print(f"sort(reverse=True)   : {copy2}")

original = [5, 2, 8, 1, 9]
ascending  = sorted(original)
descending = sorted(original, reverse=True)
print(f"\noriginal   : {original}")
print(f"sorted asc : {ascending}")
print(f"sorted desc: {descending}")

words = ["banana", "apple", "fig", "cherry", "date", "elderberry"]
by_length = sorted(words, key=len)
print(f"\nBy length: {by_length}")
by_last = sorted(words, key=lambda w: w[-1])
print(f"By last char: {by_last}")

mixed_case = ["Banana", "apple", "Cherry", "date"]
by_lower = sorted(mixed_case, key=str.lower)
print(f"Case-insensitive: {by_lower}")

students = [
    {"name": "Alice", "gpa": 3.9},
    {"name": "Bob",   "gpa": 3.5},
    {"name": "Carol", "gpa": 3.7},
]
students.sort(key=lambda s: s["gpa"], reverse=True)
print(f"\nStudents by GPA:")
for s in students:
    print(f"  {s['name']}: {s['gpa']}")
````

````python
# ============================================================
# reverse() AND copy()
# ============================================================
import copy

lst = [1, 2, 3, 4, 5]
lst.reverse()
print(f"After reverse(): {lst}")

original = [1, 2, 3, 4, 5]
rev_iter = reversed(original)
print(f"reversed()      : {list(rev_iter)}")
print(f"original intact : {original}")

lst1 = [1, [2, 3], 4]
shallow = lst1.copy()
deep    = copy.deepcopy(lst1)

lst1[0] = 99
lst1[1][0] = 999

print(f"\nOriginal       : {lst1}")
print(f"Shallow copy   : {shallow}")
print(f"Deep copy      : {deep}")

data = [(1,"z"), (2,"a"), (1,"a"), (2,"z")]
data.sort(key=lambda x: x[0])
print(f"\nStable sort: {data}")
````

## Section 5 - List Comprehensions

```python
[expression for item in iterable]
[expression for item in iterable if condition]
```

### Comparison with for-loop

```python
squares = []
for x in range(10):
    squares.append(x ** 2)

# Equivalent comprehension:
squares = [x ** 2 for x in range(10)]
```

Use comprehensions when the transformation is simple. For complex logic,
prefer a traditional loop for readability.
---

````python
# ============================================================
# LIST COMPREHENSIONS - Syntax and Patterns
# ============================================================

squares  = [x**2 for x in range(1, 11)]
print(f"Squares    : {squares}")

doubles  = [x*2 for x in [1, 2, 3, 4, 5]]
print(f"Doubles    : {doubles}")

upper_words = [w.upper() for w in ["hello", "world", "python"]]
print(f"Uppercase  : {upper_words}")

evens    = [x for x in range(1, 21) if x % 2 == 0]
print(f"\nEvens      : {evens}")

long_words = [w for w in ["hi", "hello", "world", "a", "Python"] if len(w) > 3]
print(f"Long words : {long_words}")

squared_odds = [x**2 for x in range(1, 11) if x % 2 != 0]
print(f"Squared odds: {squared_odds}")

matrix = [[i * j for j in range(1, 6)] for i in range(1, 6)]
print(f"\nMultiplication table:")
for row in matrix:
    print(" ", row)

nested = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat   = [x for row in nested for x in row]
print(f"\nFlattened  : {flat}")
````

````python
# ============================================================
# LIST COMPREHENSION PATTERNS - PRACTICAL
# ============================================================

str_nums = ["1", "2", "3", "4", "5"]
int_nums = [int(s) for s in str_nums]
print(f"String to int : {int_nums}")

students = [
    {"name": "Alice", "grade": 92},
    {"name": "Bob",   "grade": 78},
    {"name": "Carol", "grade": 85},
]
names  = [s["name"]  for s in students]
top    = [s["name"]  for s in students if s["grade"] >= 85]
print(f"Names     : {names}")
print(f"Top (>=85) : {top}")

numbers = [1, -2, 3, -4, 5, -6]
abs_vals = [x if x >= 0 else -x for x in numbers]
labels   = ["positive" if x > 0 else "negative" if x < 0 else "zero"
            for x in numbers]
print(f"\nAbsolute values: {abs_vals}")
print(f"Labels          : {labels}")

colors  = ["red", "green", "blue"]
sizes   = ["S", "M", "L"]
combos  = [f"{c}-{s}" for c in colors for s in sizes]
print(f"\nColor-Size combos ({len(combos)} total): {combos}")
````

## Section 6 - Lists as Stack and Queue

| Structure | Principle | Method | Time |
|-----------|-----------|--------|------|
| Stack (LIFO) | Push/pop from end | append() / pop() | O(1) |
| Queue (FIFO) | Enqueue end, dequeue start | append() / pop(0) | O(n) for pop(0)! |
| Queue (FIFO) | deque | append() / popleft() | O(1) |
---

````python
# ============================================================
# STACK (LIFO) USING A LIST
# ============================================================

history = []

def do_action(action):
    history.append(action)
    print(f"  Action: {action!r}  | Stack: {history}")

def undo():
    if not history:
        print("  Nothing to undo")
        return
    action = history.pop()
    print(f"  Undone: {action!r} | Stack: {history}")

print("--- Undo/Redo Stack ---")
do_action("Type Hello")
do_action("Type World")
do_action("Bold text")
do_action("Change font")
undo()
undo()
do_action("Italic text")
undo()
````

````python
# ============================================================
# QUEUE (FIFO) USING deque
# ============================================================
from collections import deque

print_queue = deque()

def submit_job(job):
    print_queue.append(job)
    print(f"  Queued  : {job!r}")

def process_next():
    if not print_queue:
        print("  Queue empty")
        return
    job = print_queue.popleft()
    print(f"  Printed : {job!r}")

print("--- Print Queue (FIFO) ---")
submit_job("Report.pdf")
submit_job("Invoice.docx")
submit_job("Photo.jpg")
process_next()
process_next()
submit_job("Spreadsheet.xlsx")
process_next()
process_next()
process_next()
````

````python
# EXERCISE 1 - Todo List Manager
todos = []

def add_todo(task, priority="medium"):
    todos.append({"task": task, "priority": priority, "done": False})
    print(f"Added: {task!r} [{priority}]")

def complete_todo(task_name):
    for t in todos:
        if t["task"] == task_name:
            t["done"] = True
            print(f"Completed: {task_name!r}")
            return
    print(f"Task not found: {task_name!r}")

def show_todos():
    pending = [t for t in todos if not t["done"]]
    done    = [t for t in todos if t["done"]]
    print(f"\n=== TODO LIST ({len(pending)} pending, {len(done)} done) ===")
    priority_order = {"high":0,"medium":1,"low":2}
    for t in sorted(todos, key=lambda x: priority_order.get(x["priority"],1)):
        status = "DONE" if t["done"] else "TODO"
        print(f"  [{status}] [{t['priority']:<6}] {t['task']}")

add_todo("Review lecture notes", "high")
add_todo("Submit assignment", "high")
add_todo("Read chapter 5", "medium")
add_todo("Exercise", "low")
complete_todo("Submit assignment")
show_todos()
````

````python
# EXERCISE 4 - Two-Stack Queue (Amortized O(1) Dequeue)
class TwoStackQueue:
    def __init__(self):
        self._inbox  = []
        self._outbox = []

    def enqueue(self, item):
        self._inbox.append(item)

    def dequeue(self):
        if not self._outbox:
            while self._inbox:
                self._outbox.append(self._inbox.pop())
        if not self._outbox:
            raise IndexError("Queue is empty")
        return self._outbox.pop()

    def __repr__(self):
        return f"Queue({self._outbox[::-1] + self._inbox})"

q = TwoStackQueue()
for i in [10, 20, 30, 40, 50]:
    q.enqueue(i)
    print(f"Enqueue {i}: {q}")

print()
for _ in range(3):
    val = q.dequeue()
    print(f"Dequeue {val}: {q}")

q.enqueue(60)
q.enqueue(70)
print(f"\nFinal: {q}")
````

````python
# CHALLENGE 1 - Merge Sort (Classic Divide and Conquer)
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left  = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i]); i += 1
        else:
            result.append(right[j]); j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

import random
data = random.sample(range(1, 101), 15)
sorted_data = merge_sort(data)
print(f"Original : {data}")
print(f"Sorted   : {sorted_data}")
print(f"Correct  : {sorted_data == sorted(data)}")
````

````python
# CHALLENGE 2 - Sliding Window Maximum
from collections import deque

def sliding_max(arr, k):
    dq = deque()
    result = []
    for i, x in enumerate(arr):
        while dq and dq[0] < i - k + 1:
            dq.popleft()
        while dq and arr[dq[-1]] <= x:
            dq.pop()
        dq.append(i)
        if i >= k - 1:
            result.append(arr[dq[0]])
    return result

arr = [1, 3, -1, -3, 5, 3, 6, 7]
k   = 3
maxima = sliding_max(arr, k)
print(f"Array  : {arr}")
print(f"Window : {k}")
print(f"Maxima : {maxima}")
````

|--------|-----------|---------|-------|
