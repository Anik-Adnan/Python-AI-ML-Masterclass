## Section 1 - Tuples: Immutable Sequences

A tuple is an ordered, immutable sequence - like a list that cannot be
changed after creation.

### When to Use Tuples vs Lists

| Use list when... | Use tuple when... |
|-------------------|---------------------|
| Contents will change | Contents are fixed by design |
| Homogeneous items | Heterogeneous fields (like a record) |
| Will be sorted/filtered/grown | Will be used as a dict key or set member |
| "collection of things" | "structured record" |

### Hashability

Tuples containing only immutable objects are hashable - usable as dict
keys and set members. Lists cannot be used this way.

```python
d = {(0, 0): "origin", (1, 0): "right"}  # tuple keys - valid
d = {[0, 0]: "origin"}                    # list keys - TypeError!
```
---

````python
# ============================================================
# TUPLES - Creation, Access, Immutability
# ============================================================

t1 = (1, 2, 3)
t2 = 1, 2, 3
t3 = (42,)
t4 = ()
t5 = tuple([1,2,3])

print(f"(1,2,3)  : {t1}, type={type(t1)}")
print(f"1,2,3    : {t2}, type={type(t2)}")
print(f"(42,)    : {t3}, type={type(t3)}")
print(f"(42)     : {type((42))}")

point = (10, 20, 30)
print(f"\npoint[0]  = {point[0]}")
print(f"point[-1] = {point[-1]}")
print(f"point[1:] = {point[1:]}")

try:
    point[0] = 99
except TypeError as e:
    print(f"\nImmutable: {e}")

a = (1, 2)
b = (3, 4)
print(f"\na + b    = {a + b}")
print(f"a * 3    = {a * 3}")
print(f"2 in a   = {2 in a}")
print(f"len(a)   = {len(a)}")
````

````python
# ============================================================
# TUPLE UNPACKING AND APPLICATIONS
# ============================================================

x, y, z = (10, 20, 30)
print(f"Unpacked: x={x}, y={y}, z={z}")

first, *rest = (1, 2, 3, 4, 5)
*head, last  = (1, 2, 3, 4, 5)
a, *middle, b = (1, 2, 3, 4, 5)
print(f"first={first}, rest={rest}")
print(f"head={head}, last={last}")
print(f"a={a}, middle={middle}, b={b}")

def statistics(numbers):
    n   = len(numbers)
    avg = sum(numbers) / n
    mn  = min(numbers)
    mx  = max(numbers)
    return n, avg, mn, mx

data = [4, 7, 2, 9, 1, 8, 5]
count, average, minimum, maximum = statistics(data)
print(f"\nStats: n={count}, avg={average:.2f}, min={minimum}, max={maximum}")

from collections import namedtuple

Point3D = namedtuple("Point3D", ["x", "y", "z"])
Student = namedtuple("Student", ["name", "gpa", "year"])

p = Point3D(1.0, 2.5, -3.0)
s = Student("Alice", 3.9, 2)
print(f"\nPoint: {p}")
print(f"  x={p.x}, y={p.y}, z={p.z}")
print(f"Student: {s}")
print(f"  name={s.name}, gpa={s.gpa}")

print(f"\np[0] == p.x: {p[0] == p.x}")
print(f"isinstance(s, tuple): {isinstance(s, tuple)}")
````

## Section 2 - Sets: Unordered Unique Collections

A set is an unordered, mutable collection of unique hashable objects.

| Property | Description |
|----------|-------------|
| Unordered | No indexing or slicing |
| Unique | Duplicates automatically removed |
| Mutable | Can add/remove elements |
| Elements hashable | Strings, ints, tuples OK; lists/dicts NOT OK |
| O(1) membership | x in s is constant time average |

### Set Algebra

| Operation | Symbol | Method |
|-----------|--------|--------|
| Union | A or B | A.union(B) |
| Intersection | A and B | A.intersection(B) |
| Difference | A minus B | A.difference(B) |
| Symmetric difference | A xor B | A.symmetric_difference(B) |
---

````python
# ============================================================
# SETS - Creation and Basic Operations
# ============================================================

s1 = {1, 2, 3, 4, 5}
s2 = set([3, 4, 5, 6, 7])
s3 = set("hello")
empty_set = set()

print(f"s1 = {s1}")
print(f"s2 = {s2}")
print(f"set(hello) = {s3}")
print(f"empty set = {empty_set}, type = {type(empty_set)}")

data_with_dups = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
unique = set(data_with_dups)
print(f"\nDuplicates removed: {data_with_dups} -> {unique}")
print(f"Back to sorted list: {sorted(unique)}")

valid_tokens = {"admin", "user", "guest", "moderator", "superadmin"}
print(f"\nadmin in valid_tokens: {'admin' in valid_tokens}")
print(f"hacker in valid_tokens: {'hacker' in valid_tokens}")

fruits = {"apple", "banana", "cherry"}
fruits.add("date")
fruits.add("apple")
print(f"\nAfter add: {fruits}")

fruits.remove("banana")
fruits.discard("kiwi")
print(f"After remove/discard: {fruits}")

popped = fruits.pop()
print(f"Popped: {popped!r}, remaining: {fruits}")
````

````python
# ============================================================
# SET ALGEBRA - Union, Intersection, Difference, etc.
# ============================================================

A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print(f"A = {A}")
print(f"B = {B}")

print(f"\nA | B  (union)         = {A | B}")
print(f"A & B  (intersection)   = {A & B}")
print(f"A - B  (difference A-B) = {A - B}")
print(f"B - A  (difference B-A) = {B - A}")
print(f"A ^ B  (symmetric diff) = {A ^ B}")

print(f"\nA <= A (subset): {A <= A}")
print(f"A <  A (proper subset): {A < A}")

class_A = {"Alice", "Bob", "Carol", "Dave", "Eve"}
class_B = {"Bob", "Carol", "Frank", "Grace", "Henry"}

both    = class_A & class_B
only_A  = class_A - class_B
only_B  = class_B - class_A
either  = class_A | class_B

print(f"\nClass A only: {sorted(only_A)}")
print(f"Class B only: {sorted(only_B)}")
print(f"In both: {sorted(both)}")
print(f"In either: {sorted(either)}")

immutable = frozenset({1, 2, 3})
d = {immutable: "frozen set as key"}
print(f"\nFrozen set key: {d}")
````

## Section 3 - Dictionaries: Key-Value Mappings

A dict is an ordered (Python 3.7+), mutable mapping from unique hashable
keys to arbitrary values. Internally implemented as a hash table,
providing O(1) average lookup, insertion, and deletion.

```python
d1 = {"name": "Alice", "age": 20}
d2 = dict(name="Alice", age=20)
d3 = dict([("name","Alice"),("age",20)])
d4 = dict.fromkeys(["a","b","c"], 0)
```
---

````python
# ============================================================
# DICTIONARIES - Creation and Access
# ============================================================

d1 = {"name": "Alice", "age": 20, "city": "Berlin"}
d2 = dict(name="Bob", age=25, city="Paris")
d3 = dict([("name","Carol"), ("age",30)])
d4 = dict.fromkeys(["monday","tuesday","wednesday"], [])

print(f"Literal     : {d1}")
print(f"dict()      : {d2}")
print(f"from pairs  : {d3}")
print(f"fromkeys    : {d4}")
print(f"type        : {type(d1)}")

student = {"name": "Alice", "gpa": 3.9, "courses": ["Math", "CS"]}

print(f"\nname: {student['name']}")
print(f"gpa : {student['gpa']}")

try:
    print(student["grade"])
except KeyError as e:
    print(f"KeyError: {e}")

print(f"\nget(name)       : {student.get('name')}")
print(f"get(grade)      : {student.get('grade')}")
print(f"get(grade,N/A)  : {student.get('grade','N/A')}")

student["year"]  = 2
student["gpa"]   = 3.95
del student["courses"]
print(f"\nModified: {student}")
````

````python
# ============================================================
# DICTIONARY METHODS - keys, values, items, pop, update
# ============================================================

inventory = {
    "apple" : {"price": 1.20, "stock": 150},
    "banana": {"price": 0.50, "stock": 300},
    "cherry": {"price": 3.00, "stock": 50},
    "date"  : {"price": 5.00, "stock": 80},
}

print("Keys  :", list(inventory.keys()))
print("Values:", [v["price"] for v in inventory.values()])

print("\nInventory:")
for fruit, info in inventory.items():
    print(f"  {fruit:<8} ${info['price']:.2f}  stock={info['stock']}")

removed = inventory.pop("date")
print(f"\nRemoved date: {removed}")
missing = inventory.pop("kiwi", None)
print(f"pop kiwi (missing): {missing}")

inventory.update({"apple": {"price": 1.50, "stock": 200}, "fig": {"price": 4.0, "stock": 20}})
print(f"\nAfter update: {list(inventory.keys())}")
print(f"Apple updated: {inventory['apple']}")

popped_item = inventory.popitem()
print(f"\npopitem() (last inserted): {popped_item}")
````

````python
# ============================================================
# setdefault() AND defaultdict
# ============================================================
from collections import defaultdict

# setdefault(key, default): get value, OR set it if missing
config = {"theme": "dark"}
font_size = config.setdefault("font_size", 14)
print(f"font_size: {font_size}, config: {config}")
theme = config.setdefault("theme", "light")
print(f"theme (already existed): {theme}, config: {config}")

# Classic use: grouping items
words = ["apple", "banana", "avocado", "cherry", "blueberry", "cantaloupe"]
groups = {}
for w in words:
    groups.setdefault(w[0], []).append(w)
print(f"\nGrouped by first letter: {groups}")

# defaultdict simplifies this pattern
grouped = defaultdict(list)
for w in words:
    grouped[w[0]].append(w)
print(f"defaultdict result: {dict(grouped)}")

# Counting with defaultdict
text = "the quick brown fox the lazy dog the fox"
counts = defaultdict(int)
for word in text.split():
    counts[word] += 1
print(f"\nWord counts: {dict(counts)}")
````

## Section 4 - Iterating Dicts and Dict Comprehensions

```python
{key_expr: value_expr for item in iterable if condition}
```
---

````python
# ============================================================
# DICT COMPREHENSIONS
# ============================================================

squares = {x: x**2 for x in range(1, 11)}
print(f"Squares dict: {squares}")

names = ["Alice", "Bob", "Carol"]
name_lengths = {name: len(name) for name in names}
print(f"Name lengths: {name_lengths}")

prices = {"apple": 1.2, "banana": 0.5, "cherry": 3.0, "date": 5.0}
expensive = {k: v for k, v in prices.items() if v > 1.0}
print(f"Expensive items: {expensive}")

# Invert a dictionary (swap keys and values)
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(f"\nOriginal: {original}")
print(f"Inverted: {inverted}")

# Combine two lists into a dict
keys = ["name", "age", "city"]
values = ["Alice", 20, "Berlin"]
combined = dict(zip(keys, values))
print(f"\nzip combined: {combined}")
````

## Section 5 - Nested Dictionaries

Dictionaries can contain other dictionaries, lists, or any combination -
a natural way to represent JSON-like structured data.
---

````python
# ============================================================
# NESTED DICTIONARIES
# ============================================================

company = {
    "name": "TechCorp",
    "founded": 2010,
    "departments": {
        "engineering": {
            "head": "Alice Johnson",
            "employees": ["Bob", "Carol", "Dave"],
            "budget": 500000,
        },
        "marketing": {
            "head": "Eve Smith",
            "employees": ["Frank", "Grace"],
            "budget": 200000,
        },
    },
}

print(f"Company: {company['name']}, founded {company['founded']}")
print(f"\nEngineering head: {company['departments']['engineering']['head']}")
print(f"Engineering team : {company['departments']['engineering']['employees']}")

# Iterate over nested structure
print("\nAll departments:")
for dept_name, dept_info in company["departments"].items():
    print(f"  {dept_name.title()}: head={dept_info['head']}, "
          f"employees={len(dept_info['employees'])}, budget=${dept_info['budget']:,}")

# Total budget across all departments
total_budget = sum(d["budget"] for d in company["departments"].values())
print(f"\nTotal budget: ${total_budget:,}")

# Add a new employee
company["departments"]["engineering"]["employees"].append("Henry")
print(f"\nAfter adding Henry: {company['departments']['engineering']['employees']}")
````

````python
# EXERCISE 1 - Tuple-based Coordinate System
points = [(0,0), (3,4), (-1,2), (5,-5), (0,7)]

def distance(p1, p2):
    return ((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2) ** 0.5

origin = (0, 0)
distances = [(p, round(distance(p, origin), 2)) for p in points]
distances.sort(key=lambda x: x[1])

print("Points sorted by distance from origin:")
for point, dist in distances:
    print(f"  {point}  distance={dist}")
````

````python
# EXERCISE 2 - Set-based Deduplication and Analysis
log_entries = ["192.168.1.1", "10.0.0.5", "192.168.1.1", "172.16.0.3",
               "10.0.0.5", "192.168.1.1", "8.8.8.8", "10.0.0.5"]

unique_ips = set(log_entries)
print(f"Total entries: {len(log_entries)}")
print(f"Unique IPs: {len(unique_ips)}")
print(f"IPs: {sorted(unique_ips)}")

from collections import Counter
counts = Counter(log_entries)
print("\nVisit counts:")
for ip, count in counts.most_common():
    print(f"  {ip:<15} {count} visits")
````

````python
# EXERCISE 3 - Dictionary-based Student Gradebook
gradebook = {
    "Alice": [88, 92, 79, 95],
    "Bob":   [72, 68, 81, 75],
    "Carol": [95, 98, 91, 99],
}

print(f"{'Student':<10} {'Average':>8} {'Highest':>8} {'Lowest':>8} Letter")
print("-" * 50)
for name, grades in gradebook.items():
    avg = sum(grades) / len(grades)
    letter = "A" if avg>=90 else "B" if avg>=80 else "C" if avg>=70 else "D" if avg>=60 else "F"
    print(f"{name:<10} {avg:>8.1f} {max(grades):>8} {min(grades):>8} {letter:>6}")

class_avg = sum(sum(g) for g in gradebook.values()) / sum(len(g) for g in gradebook.values())
print(f"\nClass-wide average: {class_avg:.2f}")
````

````python
# EXERCISE 4 - Set Operations for Recommendation
user_likes = {
    "Alice": {"Action", "Comedy", "Drama"},
    "Bob":   {"Action", "SciFi", "Horror"},
    "Carol": {"Comedy", "Romance", "Drama"},
}

def find_similar(name1, name2):
    s1, s2 = user_likes[name1], user_likes[name2]
    shared = s1 & s2
    similarity = len(shared) / len(s1 | s2) if (s1|s2) else 0
    return shared, similarity

for u1 in user_likes:
    for u2 in user_likes:
        if u1 < u2:
            shared, sim = find_similar(u1, u2)
            print(f"{u1} & {u2}: shared={shared}, similarity={sim:.2f}")
````

````python
# EXERCISE 5 - Nested Dict Builder from Flat Records
records = [
    {"dept": "Sales",  "name": "Alice", "salary": 60000},
    {"dept": "Sales",  "name": "Bob",   "salary": 55000},
    {"dept": "IT",     "name": "Carol", "salary": 75000},
    {"dept": "IT",     "name": "Dave",  "salary": 80000},
]

from collections import defaultdict
by_dept = defaultdict(list)
for r in records:
    by_dept[r["dept"]].append({"name": r["name"], "salary": r["salary"]})

for dept, employees in by_dept.items():
    total = sum(e["salary"] for e in employees)
    print(f"{dept}: {[e['name'] for e in employees]}, total salary=${total:,}")
````

````python
# CHALLENGE 1 - Simple Schema Validator
def validate_schema(data, schema):
    errors = []
    for key, expected_type in schema.items():
        if key not in data:
            errors.append(f"Missing required field: {key!r}")
        elif not isinstance(data[key], expected_type):
            errors.append(f"Field {key!r} should be {expected_type.__name__}, "
                           f"got {type(data[key]).__name__}")
    extra = set(data.keys()) - set(schema.keys())
    if extra:
        errors.append(f"Unexpected fields: {sorted(extra)}")
    return errors

user_schema = {"name": str, "age": int, "email": str, "active": bool}

test_records = [
    {"name": "Alice", "age": 25, "email": "alice@x.com", "active": True},
    {"name": "Bob", "age": "25", "email": "bob@x.com", "active": True},
    {"name": "Carol", "email": "carol@x.com", "active": False},
    {"name": "Dave", "age": 30, "email": "dave@x.com", "active": True, "extra": 1},
]

for i, record in enumerate(test_records, 1):
    errs = validate_schema(record, user_schema)
    status = "VALID" if not errs else "INVALID"
    print(f"Record {i}: {status}")
    for e in errs:
        print(f"   - {e}")
````

````python
# CHALLENGE 2 - LRU Cache using OrderedDict
from collections import OrderedDict

class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = OrderedDict()

    def get(self, key):
        if key not in self.cache:
            return -1
        self.cache.move_to_end(key)
        return self.cache[key]

    def put(self, key, value):
        if key in self.cache:
            self.cache.move_to_end(key)
        self.cache[key] = value
        if len(self.cache) > self.capacity:
            evicted_key, _ = self.cache.popitem(last=False)
            print(f"   Evicted: {evicted_key}")

cache = LRUCache(3)
cache.put("a", 1); print("put a=1  ", dict(cache.cache))
cache.put("b", 2); print("put b=2  ", dict(cache.cache))
cache.put("c", 3); print("put c=3  ", dict(cache.cache))
cache.get("a");    print("get a    ", dict(cache.cache))
cache.put("d", 4); print("put d=4  ", dict(cache.cache))
print("get b:", cache.get("b"))
````

## Summary - Day 07

| Structure | Mutable | Ordered | Duplicates | Hashable as key |
|-----------|---------|---------|------------|------------------|
| list | Yes | Yes | Yes | No |
| tuple | No | Yes | Yes | Yes (if contents immutable) |
| set | Yes | No | No | No |
| frozenset | No | No | No | Yes |
| dict | Yes | Yes (3.7+) | Keys unique | No (but values can be anything) |

| Concept | Key Point |
|---------|-----------|
| Tuple | Immutable; use for fixed records, dict keys |
| namedtuple | Tuple with named fields, readable |
| Set | Unordered unique elements; O(1) membership; full set algebra |
| Dict | Hash table; O(1) average lookup; get() avoids KeyError |
| setdefault/defaultdict | Idiomatic grouping and counting patterns |
| Dict comprehension | {k: v for ... } |
