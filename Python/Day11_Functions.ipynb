## Section 1 - Defining and Calling Functions

A **function** is a named, reusable block of code that performs a specific
task. Functions are central to **abstraction** - hiding implementation
details behind a clear interface.

### Syntax

```python
def function_name(parameter1, parameter2):
    """Optional docstring describing the function."""
    statement(s)
    return value          # Optional - omitting it returns None
```

### Terminology

| Term | Definition |
|------|-----------|
| **Parameter** | The name listed in the function definition |
| **Argument** | The actual value passed when calling the function |
| **Return value** | The value sent back to the caller via `return` |
| **Call site** | The location in code where the function is invoked |

```python
def greet(name):      # 'name' is a PARAMETER
    return f"Hello, {name}!"

greet("Alice")          # "Alice" is the ARGUMENT
```

---

````python
# ============================================================
# DEFINING AND CALLING FUNCTIONS
# ============================================================

def greet(name):
    """Return a personalised greeting for the given name."""
    return f"Hello, {name}!"

message = greet("Alice")
print(message)
print(greet("Bob"))

# ── Functions with no parameters ──
def current_year_message():
    return "Welcome to the Python course!"

print(current_year_message())

# ── Functions that return nothing (implicitly return None) ──
def log_message(text):
    print(f"[LOG] {text}")
    # No return statement -> returns None

result = log_message("System started")
print(f"log_message() returned: {result!r}")

# ── Functions calling other functions ──
def square(x):
    return x ** 2

def sum_of_squares(a, b):
    return square(a) + square(b)

print(f"\nsum_of_squares(3, 4) = {sum_of_squares(3, 4)}")    # 9 + 16 = 25

# ── Accessing docstrings ──
print(f"\ngreet.__doc__ = {greet.__doc__!r}")
help(greet)
````

## Section 2 - return: Single, Multiple, and Early Returns

### return vs print() (revisited from Day 1)

`return` sends a value back to the **caller**, allowing it to be stored,
passed to other functions, or used in expressions. `print()` only displays
text - it cannot be captured for further computation.

### Multiple Return Values

Python functions can return multiple values - actually returned as a single
**tuple**, then conventionally unpacked at the call site.

### Early Return (Guard Clause Pattern, from Day 09)

A function may contain multiple `return` statements - execution stops at the
**first one reached**.

---

````python
# ============================================================
# return - SINGLE, MULTIPLE, EARLY
# ============================================================

# ── Single return value ──
def add(a, b):
    return a + b

print(f"add(3, 4) = {add(3, 4)}")

# ── Multiple return values (implicitly a tuple) ──
def min_max_avg(numbers):
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

data = [4, 8, 15, 16, 23, 42]
lo, hi, avg = min_max_avg(data)
print(f"\nmin={lo}, max={hi}, avg={avg:.2f}")

# Without unpacking, you get the raw tuple:
result_tuple = min_max_avg(data)
print(f"Raw tuple: {result_tuple}")

# ── Early return (guard clause pattern) ──
def safe_divide(a, b):
    if b == 0:
        return None              # Early return on invalid input
    return a / b

print(f"\nsafe_divide(10, 2) = {safe_divide(10, 2)}")
print(f"safe_divide(10, 0) = {safe_divide(10, 0)}")

# ── Function with NO return: returns None implicitly ──
def print_only(x):
    print(f"Value is {x}")
    # No return

value = print_only(42)
print(f"\nReturn value of print_only(): {value!r}")

# ── return with no value also returns None (used to exit early) ──
def validate_age(age):
    if age < 0:
        print("Invalid age")
        return            # Bare return - exits with None
    print(f"Age {age} is valid")

validate_age(25)
validate_age(-5)
````

## Section 3 - Default Parameters and the Mutable Default Trap

### Default Parameter Values

```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

greet("Alice")              # Uses default: "Hello, Alice!"
greet("Bob", "Hi")           # Overrides default: "Hi, Bob!"
```

**Rule**: parameters with defaults must come **after** parameters without
defaults in the function signature.

### The Mutable Default Argument Trap (Critical!)

Default values are evaluated **once**, when the function is **defined** -
not each time it is called. Using a mutable object (list, dict, set) as a
default is one of Python's most notorious beginner pitfalls.

```python
def append_item(item, target=[]):   # DANGEROUS!
    target.append(item)
    return target

append_item(1)    # [1]
append_item(2)    # [1, 2]  <- same list reused across calls!
```

**Fix**: use `None` as the sentinel default, and create the mutable object
inside the function body.

---

````python
# ============================================================
# DEFAULT PARAMETERS
# ============================================================

def greet(name, greeting="Hello", punctuation="!"):
    return f"{greeting}, {name}{punctuation}"

print(greet("Alice"))                          # Hello, Alice!
print(greet("Bob", "Hi"))                      # Hi, Bob!
print(greet("Carol", "Hey", "?"))              # Hey, Carol?
print(greet("Dave", punctuation="..."))        # Hello, Dave...  (keyword skips middle arg)

def create_profile(name, age=18, country="Unknown", active=True):
    return {"name": name, "age": age, "country": country, "active": active}

print("\n", create_profile("Eve"))
print(create_profile("Frank", 25, "Germany"))
print(create_profile("Grace", active=False))    # Skip age/country via keyword
````

````python
# ============================================================
# THE MUTABLE DEFAULT ARGUMENT TRAP
# ============================================================

# ── BUGGY version: list default is shared across ALL calls ──
def append_item_buggy(item, target=[]):
    target.append(item)
    return target

result1 = append_item_buggy(1)
print(f"First call : {result1}")     # [1]  - looks fine so far

result2 = append_item_buggy(2)
print(f"Second call: {result2}")     # [1, 2]  - SURPRISE! Not [2]!

result3 = append_item_buggy(3)
print(f"Third call : {result3}")     # [1, 2, 3]  - The bug compounds

print(f"result1 is result2: {result1 is result2}")   # True - same object!

print()

# ── FIXED version: use None as sentinel ──
def append_item_fixed(item, target=None):
    if target is None:
        target = []          # Fresh list created EVERY call
    target.append(item)
    return target

fixed1 = append_item_fixed(1)
fixed2 = append_item_fixed(2)
fixed3 = append_item_fixed(3)
print(f"Fixed call 1: {fixed1}")    # [1]
print(f"Fixed call 2: {fixed2}")    # [2]  - correct!
print(f"Fixed call 3: {fixed3}")    # [3]  - correct!
print(f"fixed1 is fixed2: {fixed1 is fixed2}")   # False - different objects
````

## Section 4 - Argument Types: Positional, Keyword, *args, **kwargs

### Argument Passing Styles

| Style | Syntax at call site | Notes |
|-------|---------------------|-------|
| Positional | `f(1, 2)` | Matched by position |
| Keyword | `f(a=1, b=2)` | Matched by name; order-independent |
| Positional-only | parameters before `/` in def | Caller cannot use keyword form |
| Keyword-only | parameters after `*` in def | Caller MUST use keyword form |
| Variable positional | `*args` | Collects extra positional args into a tuple |
| Variable keyword | `**kwargs` | Collects extra keyword args into a dict |

### Canonical Parameter Order in a Function Definition

```python
def f(pos_only, /, normal, *args, kw_only, **kwargs):
    ...
```

---

````python
# ============================================================
# POSITIONAL VS KEYWORD ARGUMENTS
# ============================================================

def describe_pet(name, species, age):
    return f"{name} is a {age}-year-old {species}"

# Positional
print(describe_pet("Rex", "dog", 3))

# Keyword (order doesn't matter)
print(describe_pet(age=3, name="Whiskers", species="cat"))

# Mixed: positional MUST come before keyword
print(describe_pet("Tweety", age=1, species="bird"))

# ── Positional-only and keyword-only parameters (PEP 570, Python 3.8+) ──
def calculate(x, y, /, *, operation="add"):
    # x, y: positional-ONLY (before /)
    # operation: keyword-ONLY (after *)
    if operation == "add": return x + y
    if operation == "sub": return x - y
    raise ValueError(f"Unknown operation: {operation}")

print(f"\ncalculate(5, 3) = {calculate(5, 3)}")
print(f"calculate(5, 3, operation='sub') = {calculate(5, 3, operation='sub')}")

try:
    calculate(x=5, y=3)     # Fails! x, y are positional-only
except TypeError as e:
    print(f"\nTypeError (positional-only violated): {e}")

try:
    calculate(5, 3, "sub")   # Fails! operation is keyword-only
except TypeError as e:
    print(f"TypeError (keyword-only violated): {e}")
````

````python
# ============================================================
# *args AND **kwargs
# ============================================================

# ── *args: collects extra POSITIONAL arguments into a tuple ──
def total(*args):
    print(f"  args = {args}  (type: {type(args).__name__})")
    return sum(args)

print(f"total(1,2,3)       = {total(1, 2, 3)}")
print(f"total(1,2,3,4,5)   = {total(1, 2, 3, 4, 5)}")
print(f"total()            = {total()}")

# ── **kwargs: collects extra KEYWORD arguments into a dict ──
def build_profile(**kwargs):
    print(f"  kwargs = {kwargs}  (type: {type(kwargs).__name__})")
    return kwargs

print()
profile = build_profile(name="Alice", age=20, city="Berlin")
print(f"Profile: {profile}")

# ── Combining normal params, *args, and **kwargs ──
def flexible_function(required, *args, optional="default", **kwargs):
    print(f"  required = {required}")
    print(f"  args     = {args}")
    print(f"  optional = {optional}")
    print(f"  kwargs   = {kwargs}")

print()
flexible_function(1, 2, 3, optional="custom", extra1="a", extra2="b")

# ── Unpacking with * and ** at the CALL site (the reverse operation) ──
def add3(a, b, c):
    return a + b + c

numbers = [1, 2, 3]
print(f"\nadd3(*numbers) = {add3(*numbers)}")   # Unpacks list into positional args

kwargs_dict = {"a": 10, "b": 20, "c": 30}
print(f"add3(**kwargs_dict) = {add3(**kwargs_dict)}")   # Unpacks dict into keyword args
````

## Section 5 - Scope: The LEGB Rule

Python resolves a name using the **LEGB** rule, searching in this order:

| Scope | Description |
|-------|--------------|
| **L**ocal | Names assigned within the current function |
| **E**nclosing | Names in any enclosing function (for nested functions/closures) |
| **G**lobal | Names assigned at the module (top) level |
| **B**uilt-in | Names provided by Python itself (`print`, `len`, etc.) |

### The `global` and `nonlocal` Keywords

By default, assigning to a name inside a function creates a **new local
variable** - it does not modify an outer-scope variable of the same name.
`global` and `nonlocal` override this behaviour explicitly.

---

````python
# ============================================================
# SCOPE - LEGB RULE
# ============================================================

x = "global x"     # Global scope

def outer():
    y = "enclosing y"    # Enclosing scope (relative to inner())
    def inner():
        z = "local z"     # Local scope
        print(f"  Inside inner(): z={z!r}, y={y!r}, x={x!r}")  # Can read all 3
    inner()

outer()

print()

# ── Assignment creates a LOCAL variable by default (does not affect global) ──
counter = 0

def increment_broken():
    counter = counter + 1   # UnboundLocalError! Python sees assignment -> treats as local
    return counter

try:
    increment_broken()
except UnboundLocalError as e:
    print(f"UnboundLocalError: {e}")

# ── Fix using 'global' ──
def increment_fixed():
    global counter
    counter = counter + 1
    return counter

print(f"\nincrement_fixed() x3: {increment_fixed()}, {increment_fixed()}, {increment_fixed()}")
print(f"Final global counter: {counter}")
````

````python
# ============================================================
# nonlocal - MODIFYING ENCLOSING SCOPE
# ============================================================

def make_counter():
    count = 0           # Enclosing scope variable

    def increment():
        nonlocal count   # Without this, 'count += 1' would raise UnboundLocalError
        count += 1
        return count

    return increment     # Return the inner function itself (a CLOSURE - Day 12)

counter1 = make_counter()
counter2 = make_counter()   # Independent counter - separate enclosing scope

print(f"counter1(): {counter1()}")   # 1
print(f"counter1(): {counter1()}")   # 2
print(f"counter1(): {counter1()}")   # 3
print(f"counter2(): {counter2()}")   # 1  - independent from counter1!

# ── Demonstrating WITHOUT nonlocal (broken version) ──
def make_counter_broken():
    count = 0
    def increment():
        count += 1   # UnboundLocalError without nonlocal
        return count
    return increment

broken_counter = make_counter_broken()
try:
    broken_counter()
except UnboundLocalError as e:
    print(f"\nWithout nonlocal: UnboundLocalError - {e}")
````

## Section 6 - Docstrings and Function Documentation

A **docstring** is a string literal placed as the first statement inside a
function, class, or module. It is accessible via `help()` and `.__doc__`,
and is the standard way to document Python code.

### PEP 257 Docstring Conventions

```python
def calculate_area(radius):
    """
    Calculate the area of a circle given its radius.

    Args:
        radius (float): The radius of the circle. Must be non-negative.

    Returns:
        float: The area of the circle.

    Raises:
        ValueError: If radius is negative.
    """
    if radius < 0:
        raise ValueError("Radius cannot be negative")
    return 3.14159 * radius ** 2
```

---

````python
# ============================================================
# DOCSTRINGS - Documentation Convention
# ============================================================

def calculate_area(radius):
    """
    Calculate the area of a circle given its radius.

    Args:
        radius (float): The radius of the circle. Must be non-negative.

    Returns:
        float: The area of the circle.

    Raises:
        ValueError: If radius is negative.
    """
    if radius < 0:
        raise ValueError("Radius cannot be negative")
    return 3.14159 * radius ** 2

print(calculate_area(5))
print(f"\nDocstring:\n{calculate_area.__doc__}")

# ── Type hints (preview - formally optional in Python, but common in industry) ──
def add_typed(a: int, b: int) -> int:
    """Add two integers and return the result."""
    return a + b

print(f"\nadd_typed(3, 4) = {add_typed(3, 4)}")
print(f"Type hints (not enforced at runtime): {add_typed.__annotations__}")
````

````python
# EXERCISE 1 - Temperature Converter Functions
def celsius_to_fahrenheit(c):
    """Convert Celsius to Fahrenheit."""
    return c * 9/5 + 32

def fahrenheit_to_celsius(f):
    """Convert Fahrenheit to Celsius."""
    return (f - 32) * 5/9

def kelvin_to_celsius(k):
    """Convert Kelvin to Celsius."""
    return k - 273.15

for c in [0, 20, 37, 100]:
    print(f"  {c}C = {celsius_to_fahrenheit(c):.1f}F")
print()
for f in [32, 68, 98.6, 212]:
    print(f"  {f}F = {fahrenheit_to_celsius(f):.1f}C")
````

````python
# EXERCISE 2 - Statistics Function Suite
def mean(numbers):
    return sum(numbers) / len(numbers)

def variance(numbers):
    m = mean(numbers)
    return sum((x - m) ** 2 for x in numbers) / len(numbers)

def std_dev(numbers):
    return variance(numbers) ** 0.5

def summarize(numbers):
    return {
        "count": len(numbers),
        "mean": round(mean(numbers), 2),
        "variance": round(variance(numbers), 2),
        "std_dev": round(std_dev(numbers), 2),
        "min": min(numbers),
        "max": max(numbers),
    }

data = [23, 45, 12, 67, 34, 89, 21, 56]
print(summarize(data))
````

````python
# EXERCISE 3 - Flexible Logger using *args and **kwargs
def log(level, *messages, **metadata):
    msg = " ".join(str(m) for m in messages)
    meta_str = ", ".join(f"{k}={v}" for k, v in metadata.items())
    output = f"[{level.upper()}] {msg}"
    if meta_str:
        output += f"  ({meta_str})"
    print(output)

log("info", "Server started")
log("error", "Connection", "failed", code=500, retry=True)
log("debug", "x =", 42, "y =", 17, source="module_a")
````

````python
# EXERCISE 4 - Default Argument Practice: Shopping Cart Item
def create_item(name, price, quantity=1, discount=0.0):
    subtotal = price * quantity
    total = subtotal * (1 - discount)
    return {"name": name, "price": price, "quantity": quantity,
            "discount": discount, "total": round(total, 2)}

items = [
    create_item("Book", 15.99, 2),
    create_item("Pen", 1.50, 10, discount=0.10),
    create_item("Laptop", 999.99),
]
grand_total = 0
for item in items:
    print(f"  {item['name']:<10} qty={item['quantity']:<3} total=${item['total']:.2f}")
    grand_total += item["total"]
print(f"\nGrand total: ${grand_total:.2f}")
````

````python
# EXERCISE 5 - Recursive Scope Demonstration
def outer_function():
    message = "outer"

    def middle_function():
        message = "middle"   # Shadows outer's message (new local in this scope)

        def inner_function():
            nonlocal message   # Refers to middle_function's message
            message = "modified by inner"
            return message

        result = inner_function()
        return result, message   # message reflects inner's modification

    return middle_function()

result = outer_function()
print(f"Result: {result}")
print("Demonstrates: nonlocal binds to the NEAREST enclosing scope, not global")
````

````python
# CHALLENGE 1 - Memoised Fibonacci (deliberate, controlled use of mutable default)
# NOTE: This is one of the RARE cases where a mutable default is used on
# purpose, because we WANT the cache to persist across calls. This is safe
# specifically because we never reassign the default - we only mutate it.
def fib_memo(n, cache={0: 0, 1: 1}):
    if n in cache:
        return cache[n]
    cache[n] = fib_memo(n - 1, cache) + fib_memo(n - 2, cache)
    return cache[n]

for n in [10, 20, 30, 40, 50]:
    print(f"  fib({n}) = {fib_memo(n)}")

print(f"\nCache now contains {len(fib_memo.__defaults__[0])} entries (persisted across calls)")
````

````python
# CHALLENGE 2 - Generic Validator Builder
def make_validator(**rules):
    """
    Build a validator function from keyword rules.
    Supported rules: min_value, max_value, min_length, max_length, required_type
    """
    def validate(value):
        errors = []
        if "required_type" in rules and not isinstance(value, rules["required_type"]):
            errors.append(f"Expected type {rules['required_type'].__name__}, got {type(value).__name__}")
            return errors   # Skip further checks if type is wrong
        if "min_value" in rules and value < rules["min_value"]:
            errors.append(f"Value {value} is below minimum {rules['min_value']}")
        if "max_value" in rules and value > rules["max_value"]:
            errors.append(f"Value {value} exceeds maximum {rules['max_value']}")
        if "min_length" in rules and len(value) < rules["min_length"]:
            errors.append(f"Length {len(value)} is below minimum {rules['min_length']}")
        if "max_length" in rules and len(value) > rules["max_length"]:
            errors.append(f"Length {len(value)} exceeds maximum {rules['max_length']}")
        return errors

    return validate

age_validator = make_validator(required_type=int, min_value=0, max_value=120)
name_validator = make_validator(required_type=str, min_length=2, max_length=50)

test_ages = [25, -5, 150, "twenty"]
for age in test_ages:
    errs = age_validator(age)
    print(f"  age={age!r}: {'VALID' if not errs else errs}")

test_names = ["Alice", "A", "X" * 60]
for name in test_names:
    errs = name_validator(name)
    print(f"  name={name[:20]!r}{'...' if len(name)>20 else ''}: {'VALID' if not errs else errs}")
````
