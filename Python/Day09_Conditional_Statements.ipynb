## Section 1 - The if Statement

### Syntax

```python
if condition:
    statement(s)        # Executes only if condition is truthy
```

The condition can be any expression - it is evaluated for truthiness using
the rules from Day 02 (falsy values: 0, 0.0, "", [], {}, None, False).

### if-else

```python
if condition:
    statement(s)         # Runs if condition is True
else:
    statement(s)         # Runs if condition is False
```

### if-elif-else Chain

```python
if condition1:
    ...
elif condition2:
    ...
elif condition3:
    ...
else:
    ...                   # Runs only if ALL preceding conditions are False
```

Conditions are checked **top to bottom**, and **only the first matching
branch executes** - even if a later condition would also be true.

---

````python
# ============================================================
# BASIC if, if-else, if-elif-else
# ============================================================

# ── Simple if ──
temperature = 35
if temperature > 30:
    print("It is hot today.")

# ── if-else ──
age = 16
if age >= 18:
    print("You may vote.")
else:
    print("You may not vote yet.")

# ── if-elif-else chain ──
score = 73
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"\nScore {score} -> Grade {grade}")

# ── Only the FIRST matching branch executes ──
x = 15
if x > 5:
    print("\nx > 5")        # This prints
elif x > 10:
    print("x > 10")        # This does NOT print, even though it is true!
elif x > 0:
    print("x > 0")         # This does NOT print either
````

## Section 2 - Nested Conditionals

A conditional statement can contain another conditional statement inside its
body. This is necessary when a decision depends on **multiple independent
factors** evaluated in sequence.

### When to Nest vs When to Chain

| Use nesting when... | Use elif chain when... |
|----------------------|--------------------------|
| Conditions are hierarchical (outer gates inner) | Conditions are mutually exclusive alternatives |
| Inner logic only applies within outer context | All conditions test the same variable/range |

```python
# Nesting (hierarchical):
if user_logged_in:
    if user_is_admin:
        ...

# Chaining (alternatives):
if   score >= 90: grade = "A"
elif score >= 80: grade = "B"
```

---

````python
# ============================================================
# NESTED CONDITIONALS
# ============================================================

def classify_triangle(a, b, c):
    sides = sorted([a, b, c])
    if sides[0] + sides[1] <= sides[2]:
        return "Not a valid triangle"

    if a == b == c:
        return "Equilateral"
    else:
        if a == b or b == c or a == c:
            return "Isosceles"
        else:
            return "Scalene"

triangles = [(3, 3, 3), (3, 3, 5), (3, 4, 5), (1, 1, 5)]
for t in triangles:
    print(f"  Sides {t}: {classify_triangle(*t)}")

print()

# ── Login authorization example (genuinely hierarchical) ──
def check_access(logged_in, role, resource_locked):
    if logged_in:
        if role == "admin":
            return "Full access granted"
        elif role == "editor":
            if not resource_locked:
                return "Edit access granted"
            else:
                return "Resource is locked - read-only"
        else:
            return "Read-only access granted"
    else:
        return "Access denied - please log in"

cases = [
    (True, "admin", True),
    (True, "editor", False),
    (True, "editor", True),
    (True, "viewer", False),
    (False, "admin", False),
]
for logged_in, role, locked in cases:
    print(f"  logged_in={logged_in}, role={role}, locked={locked}: "
          f"{check_access(logged_in, role, locked)}")
````

## Section 3 - The Ternary (Conditional) Expression

Python provides a single-line conditional expression for choosing between two
values:

```python
value = expr_if_true if condition else expr_if_false
```

This is **not** a replacement for full if-elif-else logic - it is appropriate
only for simple value selection, not for executing multiple statements.

```python
# Equivalent:
if age >= 18:
    status = "adult"
else:
    status = "minor"

status = "adult" if age >= 18 else "minor"   # Same result, one line
```

Ternary expressions can be chained, but readability degrades quickly beyond
two branches.

---

````python
# ============================================================
# TERNARY (CONDITIONAL) EXPRESSIONS
# ============================================================

age = 20
status = "adult" if age >= 18 else "minor"
print(f"Age {age}: {status}")

# ── Used inside other expressions ──
numbers = [-5, 3, -8, 12, 0, -1, 7]
abs_values = [n if n >= 0 else -n for n in numbers]
print(f"\nAbsolute values: {abs_values}")

# ── Used inside f-strings ──
score = 55
print(f"\nResult: {'PASS' if score >= 60 else 'FAIL'}")

# ── Chained ternary (use sparingly!) ──
def grade_letter(score):
    return ("A" if score >= 90 else
            "B" if score >= 80 else
            "C" if score >= 70 else
            "D" if score >= 60 else "F")

for s in [95, 82, 71, 64, 40]:
    print(f"  Score {s}: Grade {grade_letter(s)}")

# ── When NOT to use ternary: multi-statement logic needs full if-else ──
def process(value):
    # This CANNOT be a ternary - multiple statements per branch
    if value > 0:
        print("Processing positive value")
        result = value * 2
    else:
        print("Processing non-positive value")
        result = 0
    return result

print(f"\nprocess(5)  = {process(5)}")
print(f"process(-3) = {process(-3)}")
````

## Section 4 - Structural Pattern Matching (match-case, Python 3.10+)

Python 3.10 introduced `match-case`, a powerful alternative to long
`if-elif` chains, especially useful for matching against literal values,
types, and structured data (tuples, lists, dicts).

### Basic Syntax

```python
match subject:
    case pattern1:
        ...
    case pattern2:
        ...
    case _:            # Wildcard - matches anything (like 'default')
        ...
```

### Pattern Types

| Pattern | Example | Matches |
|---------|---------|---------|
| Literal | `case 404:` | Exact value |
| Or-pattern | `case 404 \| 500:` | Either value |
| Capture | `case x:` | Anything, binds to `x` |
| Wildcard | `case _:` | Anything (no binding) |
| Sequence | `case [x, y]:` | A 2-element sequence |
| Mapping | `case {"key": v}:` | Dict containing that key |
| Guard | `case x if x > 0:` | Pattern AND extra condition |

---

````python
# ============================================================
# match-case (STRUCTURAL PATTERN MATCHING, Python 3.10+)
# ============================================================
import sys
print(f"Python version: {sys.version_info.major}.{sys.version_info.minor}")

def http_status_message(code):
    match code:
        case 200:
            return "OK"
        case 201:
            return "Created"
        case 400 | 401 | 403:
            return "Client Error (Bad Request / Unauthorized / Forbidden)"
        case 404:
            return "Not Found"
        case 500 | 502 | 503:
            return "Server Error"
        case _ if 200 <= code < 300:
            return "Success (other 2xx)"
        case _:
            return "Unknown Status"

for code in [200, 201, 404, 401, 500, 299, 999]:
    print(f"  {code}: {http_status_message(code)}")

print()

# ── Matching sequences (destructuring) ──
def describe_point(point):
    match point:
        case (0, 0):
            return "Origin"
        case (0, y):
            return f"On the Y-axis at y={y}"
        case (x, 0):
            return f"On the X-axis at x={x}"
        case (x, y) if x == y:
            return f"On the line y=x at ({x},{y})"
        case (x, y):
            return f"Point at ({x}, {y})"
        case _:
            return "Not a 2D point"

points = [(0,0), (0,5), (3,0), (4,4), (2,7)]
for p in points:
    print(f"  {p}: {describe_point(p)}")
````

````python
# ============================================================
# match-case WITH DICTIONARIES AND TYPES
# ============================================================

def handle_event(event):
    match event:
        case {"type": "click", "x": x, "y": y}:
            return f"Click at ({x}, {y})"
        case {"type": "keypress", "key": key}:
            return f"Key pressed: {key!r}"
        case {"type": "scroll", "delta": delta} if delta > 0:
            return "Scrolled up"
        case {"type": "scroll", "delta": delta}:
            return "Scrolled down"
        case _:
            return "Unknown event"

events = [
    {"type": "click", "x": 10, "y": 20},
    {"type": "keypress", "key": "Enter"},
    {"type": "scroll", "delta": 5},
    {"type": "scroll", "delta": -3},
    {"type": "resize"},
]
for e in events:
    print(f"  {e} -> {handle_event(e)}")

print()

# ── Matching on type ──
def describe_value(val):
    match val:
        case int() if val < 0:
            return "negative integer"
        case int():
            return "non-negative integer"
        case float():
            return "float"
        case str() if len(val) == 0:
            return "empty string"
        case str():
            return "string"
        case list() | tuple():
            return f"sequence of length {len(val)}"
        case None:
            return "None"
        case _:
            return "unknown type"

for v in [42, -5, 3.14, "hi", "", [1,2,3], (1,2), None]:
    print(f"  {v!r:<15} -> {describe_value(v)}")
````

## Section 5 - Guard Clauses: Reducing Nesting

Deeply nested `if` blocks are difficult to read. A **guard clause** is an
early `return` (or `continue`/`break`) at the top of a function that handles
edge cases immediately, allowing the rest of the function to assume valid
state - this flattens the structure.

```python
# Deeply nested (harder to read):
def process(data):
    if data is not None:
        if len(data) > 0:
            if all(isinstance(x, int) for x in data):
                return sum(data) / len(data)
            else:
                return None
        else:
            return None
    else:
        return None

# Refactored with guard clauses (flat, readable):
def process(data):
    if data is None:
        return None
    if len(data) == 0:
        return None
    if not all(isinstance(x, int) for x in data):
        return None
    return sum(data) / len(data)
```

---

````python
# ============================================================
# GUARD CLAUSES - Before and After
# ============================================================

# ── BEFORE: deeply nested ──
def process_nested(data):
    if data is not None:
        if len(data) > 0:
            if all(isinstance(x, (int, float)) for x in data):
                if all(x >= 0 for x in data):
                    return sum(data) / len(data)
                else:
                    return None
            else:
                return None
        else:
            return None
    else:
        return None

# ── AFTER: flattened with guard clauses ──
def process_flat(data):
    if data is None:
        return None
    if len(data) == 0:
        return None
    if not all(isinstance(x, (int, float)) for x in data):
        return None
    if not all(x >= 0 for x in data):
        return None
    return sum(data) / len(data)

test_cases = [
    [1, 2, 3, 4],
    [],
    None,
    [1, "two", 3],
    [1, -2, 3],
]
print("Comparing nested vs flat implementations:")
for case in test_cases:
    r1 = process_nested(case)
    r2 = process_flat(case)
    match = "MATCH" if r1 == r2 else "MISMATCH"
    print(f"  {str(case):<20} nested={r1}, flat={r2}  [{match}]")
````

## Section 6 - Common Conditional Mistakes

| Mistake | Example | Fix |
|---------|---------|-----|
| Using `=` instead of `==` | `if x = 5:` | `if x == 5:` (Python raises SyntaxError, unlike C) |
| Redundant comparison to True/False | `if flag == True:` | `if flag:` |
| Comparing floats with `==` | `if x == 0.3:` | `if math.isclose(x, 0.3):` |
| Forgetting `elif` (using separate `if`s) | Multiple independent `if`s checked even after match | Use `elif` for mutually exclusive cases |
| Confusing `and`/`or` logic | `if 0 <= x and x <= 10` works but `0 <= x <= 10` is clearer | Use chained comparison |

---

````python
# ============================================================
# COMMON MISTAKES - Demonstrated and Corrected
# ============================================================

# ── Mistake 1: redundant True/False comparison ──
flag = True
if flag == True:        # Works, but redundant
    print("Mistake 1 (works but not idiomatic)")
if flag:                 # Idiomatic
    print("Correct idiom")

# ── Mistake 2: float equality ──
import math
x = 0.1 + 0.2
if x == 0.3:
    print("Floats are exactly equal")
else:
    print("Floats are NOT exactly equal (==): demonstrates the float pitfall")
if math.isclose(x, 0.3):
    print("isclose() correctly identifies them as equal")

# ── Mistake 3: separate if statements instead of elif (performance/logic bug) ──
def grade_wrong(score):
    result = []
    if score >= 90: result.append("A")
    if score >= 80: result.append("B")   # Also true for a 95! Logic error.
    if score >= 70: result.append("C")
    return result

def grade_correct(score):
    if score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    return "F"

print(f"\nscore=95 with separate ifs : {grade_wrong(95)}")   # Bug: multiple grades!
print(f"score=95 with elif chain   : {grade_correct(95)}")   # Correct: just A

# ── Mistake 4: verbose range check vs chained comparison ──
val = 7
verbose = 0 <= val and val <= 10
chained = 0 <= val <= 10
print(f"\nverbose check: {verbose}, chained check: {chained}")  # Same result, chained is cleaner
````

````python
# EXERCISE 2 - Leap Year Checker
def is_leap_year(year):
    if year % 4 != 0:
        return False
    elif year % 100 != 0:
        return True
    elif year % 400 != 0:
        return False
    else:
        return True

for year in [2000, 1900, 2024, 2023, 2400, 1800]:
    print(f"  {year}: {'Leap' if is_leap_year(year) else 'Not leap'}")
````

````python
# EXERCISE 3 - Simple Calculator using match-case
def calculate(a, op, b):
    match op:
        case "+": return a + b
        case "-": return a - b
        case "*": return a * b
        case "/":
            if b == 0:
                return "Error: division by zero"
            return a / b
        case "**": return a ** b
        case _: return f"Unknown operator: {op!r}"

operations = [(10, "+", 5), (10, "-", 5), (10, "*", 5), (10, "/", 0), (2, "**", 8), (10, "%", 5)]
for a, op, b in operations:
    print(f"  {a} {op} {b} = {calculate(a, op, b)}")
````

````python
# EXERCISE 4 - BMI Classifier with Age-adjusted Thresholds
def classify_bmi(weight_kg, height_m, age):
    bmi = weight_kg / height_m ** 2
    if age < 18:
        category = "Use pediatric growth charts (not adult BMI)"
    elif bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal weight"
    elif bmi < 30:
        category = "Overweight"
    else:
        category = "Obese"
    return round(bmi, 1), category

people = [(70, 1.75, 30), (45, 1.60, 16), (95, 1.70, 45), (55, 1.65, 25)]
for w, h, age in people:
    bmi, cat = classify_bmi(w, h, age)
    print(f"  weight={w}kg height={h}m age={age}: BMI={bmi}, {cat}")
````

````python
# EXERCISE 5 - Rock Paper Scissors Winner Logic
def rps_winner(p1, p2):
    if p1 == p2:
        return "Tie"
    beats = {"rock": "scissors", "scissors": "paper", "paper": "rock"}
    if beats.get(p1) == p2:
        return "Player 1 wins"
    else:
        return "Player 2 wins"

matchups = [("rock","scissors"), ("paper","rock"), ("scissors","scissors"), ("paper","scissors")]
for p1, p2 in matchups:
    print(f"  {p1} vs {p2}: {rps_winner(p1, p2)}")
````

````python
# CHALLENGE 1 - Traffic Light State Machine
def next_light_state(current, has_pedestrian_request=False):
    match current:
        case "red":
            return "green"
        case "green":
            if has_pedestrian_request:
                return "yellow"
            return "green"
        case "yellow":
            return "red"
        case _:
            raise ValueError(f"Invalid state: {current!r}")

state = "red"
print(f"Start: {state}")
for step in range(6):
    pedestrian = (step == 2)
    state = next_light_state(state, pedestrian)
    print(f"  Step {step+1} (pedestrian_request={pedestrian}): -> {state}")
````

````python
# CHALLENGE 2 - Progressive Tax Bracket Calculator
def calculate_tax(income):
    brackets = [
        (0,      10000,  0.10),
        (10000,  40000,  0.12),
        (40000,  85000,  0.22),
        (85000, 160000,  0.24),
        (160000, float("inf"), 0.32),
    ]
    total_tax = 0.0
    breakdown = []
    for lower, upper, rate in brackets:
        if income <= lower:
            break
        taxable_in_bracket = min(income, upper) - lower
        tax_in_bracket = taxable_in_bracket * rate
        total_tax += tax_in_bracket
        breakdown.append((lower, min(income, upper), rate, tax_in_bracket))
        if income <= upper:
            break
    return total_tax, breakdown

for income in [5000, 35000, 75000, 200000]:
    tax, breakdown = calculate_tax(income)
    print(f"\nIncome: ${income:,}")
    for lower, upper, rate, amount in breakdown:
        print(f"  ${lower:>7,.0f} - ${upper:>7,.0f} @ {rate*100:>4.0f}% = ${amount:>9,.2f}")
    print(f"  Total tax: ${tax:,.2f}  (effective rate: {tax/income*100:.1f}%)")
````
