## 🔷 Section 1 — Variables and Python's Memory Model

### What Is a Variable in Python?

In languages like C or Java, a variable is a **named memory box** that stores a value directly.
In Python, a variable is a **label (reference) that points to an object** in memory.

```
C/C++  :  int x = 5;   → [memory cell named x] holds value 5
Python :  x = 5        → object int(5) exists in memory; name 'x' points to it
```

This distinction has profound consequences:

```python
x = [1, 2, 3]
y = x            # y is NOT a copy — it points to the SAME list object
y.append(4)
print(x)         # [1, 2, 3, 4] — x is affected because x and y are the same object
```

The `=` operator **binds a name to an object**. It does not copy the object.
We will explore this fully when we cover lists (Day 6) and mutation.

### PEP 8 Naming Rules

| Rule | Valid Examples | Invalid Examples |
|------|---------------|-----------------|
| Start with letter or `_` | `name`, `_private` | `1name`, `@var` |
| Only letters, digits, `_` | `my_var2` | `my-var`, `my var` |
| Not a reserved keyword | `for_loop` | `for`, `class`, `if` |
| snake_case for variables | `student_age` | `StudentAge` (PascalCase — for classes) |
| ALL_CAPS for constants | `MAX_SIZE = 100` | — |

---

````python
# ============================================================
# VARIABLES — Assignment, Rebinding, Multiple References
# ============================================================

# ── Basic assignment: name bound to an object ──
x = 42
print(f"x = {x}, type = {type(x)}, id = {id(x)}")

# ── Rebinding: x now points to a different object ──
x = "hello"
print(f"x = {x!r}, type = {type(x)}, id = {id(x)}")
# The integer 42 is now unreferenced → eligible for garbage collection

# ── Two names, same object ──
a = [1, 2, 3]
b = a                      # b and a reference the SAME list
print(f"a is b: {a is b}") # True — 'is' checks identity (same id)
b.append(4)
print(f"a after b.append(4): {a}")  # [1, 2, 3, 4] — mutation visible via a

# ── Contrast with immutable integers: Python caches small ints (-5 to 256) ──
p = 100
q = 100
print(f"p is q (value 100): {p is q}")   # True — cached

r = 1000
s = 1000
print(f"r is s (value 1000): {r is s}")  # May be False — not cached
````

````python
# ============================================================
# NAMING RULES AND RESERVED KEYWORDS
# ============================================================
import keyword

# ── All Python reserved keywords ──
kws = keyword.kwlist
import sys
print(f"Number of reserved keywords in Python {sys.version_info.major}.{sys.version_info.minor}: {len(kws)}")
print("Keywords:", kws)

# ── Legal variable names ──
student_name  = "Alice"      # snake_case — preferred
_private_var  = 42           # leading underscore — conventional "private"
CamelCase     = "classes"    # PascalCase — used for class names
__dunder__    = "special"    # double underscores — used by Python internally
counter2025   = 0            # digits allowed (not at start)

# ── Illegal names (commented out — would raise SyntaxError) ──
# 2fast = "no"     ← starts with digit
# my-var = 1       ← hyphen not allowed
# for = 10         ← 'for' is a keyword

# ── Common pitfall: shadowing built-ins ──
# Avoid naming your variables after built-in functions!
# list = [1, 2, 3]   ← shadows the built-in list type — bad practice!
# print = 5          ← would prevent calling print() — very bad!

print("\nAll naming checks passed.")
````

### 1.2 Multiple Assignment and Augmented Operators

Python supports several compact assignment patterns that make code more readable
and idiomatic.

#### Multiple Assignment (Same Object)
```python
a = b = c = 0      # All three names point to the same int object 0
```

#### Tuple Unpacking (Parallel Assignment)
```python
x, y = 10, 20      # x=10, y=20 — right side is an implicit tuple
a, b, c = 1, 2, 3  # Three names, three values
first, *rest = [1, 2, 3, 4]   # first=1, rest=[2,3,4] — extended unpacking
```

#### Variable Swap (Pythonic idiom)
```python
x, y = y, x    # No temp variable needed — Python evaluates right side fully first
```

#### Augmented Assignment Operators

| Operator | Equivalent | Example |
|----------|-----------|---------|
| `+=` | `x = x + n` | `x += 1` |
| `-=` | `x = x - n` | `x -= 5` |
| `*=` | `x = x * n` | `x *= 2` |
| `/=` | `x = x / n` | `x /= 4` |
| `//=` | `x = x // n` | `x //= 3` |
| `**=` | `x = x ** n` | `x **= 2` |
| `%=` | `x = x % n` | `x %= 10` |

---

````python
# ============================================================
# MULTIPLE ASSIGNMENT AND TUPLE UNPACKING
# ============================================================

# ── Multiple names → same object ──
a = b = c = 0
print(f"a={a}, b={b}, c={c}")
print(f"Same object? a is b: {a is b}, b is c: {b is c}")  # True (int cache)

# ── Tuple unpacking (parallel assignment) ──
x, y = 10, 20
print(f"x={x}, y={y}")

# ── Swap without a temporary variable ──
print(f"Before swap: x={x}, y={y}")
x, y = y, x      # Python evaluates (y, x) as tuple (20, 10) first, then unpacks
print(f"After swap:  x={x}, y={y}")

# ── Extended unpacking with * ──
first, *middle, last = [1, 2, 3, 4, 5]
print(f"first={first}, middle={middle}, last={last}")

# ── Unpacking from a function return (preview of Day 11) ──
def min_max(lst): return min(lst), max(lst)
lo, hi = min_max([3, 1, 7, 2, 9])
print(f"min={lo}, max={hi}")
````

````python
# ============================================================
# AUGMENTED ASSIGNMENT OPERATORS
# ============================================================
# These combine an operation and assignment in one step.
# x += n  is equivalent to  x = x + n
# IMPORTANT: for mutable objects (lists, etc.) += may mutate in place.

n = 10
print(f"Initial n = {n}")

n += 5;  print(f"After += 5  : {n}")   # 15
n -= 3;  print(f"After -= 3  : {n}")   # 12
n *= 2;  print(f"After *= 2  : {n}")   # 24
n //= 5; print(f"After //= 5 : {n}")   # 4
n **= 3; print(f"After **= 3 : {n}")   # 64
n %= 10; print(f"After %= 10 : {n}")   # 4
n /= 2;  print(f"After /= 2  : {n}")   # 2.0  (float!)

# ── Practical: building a running total ──
total = 0
prices = [9.99, 14.50, 3.25, 22.00]
for price in prices:          # (Day 10: for-loop preview)
    total += price
print(f"\nCart total: ${total:.2f}")

# ── += on strings (concatenation) ──
message = "Hello"
message += ", World"
message += "!"
print(message)    # Hello, World!
````

## 🔷 Section 2 — Python's Primitive Data Types

Python has **five fundamental built-in types** for scalar values:

| Type | Python name | Example literals | Mutable? |
|------|------------|-----------------|---------|
| Integer | `int` | `0`, `42`, `-17`, `2**64` | No |
| Floating-point | `float` | `3.14`, `-0.5`, `1e8` | No |
| String | `str` | `"hello"`, `'world'`, `"""multi"""` | No |
| Boolean | `bool` | `True`, `False` | No |
| Null | `NoneType` | `None` | No (singleton) |

All of these are **immutable** — once created, their value cannot change.
(When you do `x = x + 1`, Python creates a *new* int object; it does not modify the existing one.)

---

````python
# ============================================================
# INTEGER TYPE — int
# ============================================================
# Python integers have UNLIMITED precision (arbitrary-precision arithmetic).
# There is no int overflow in Python (unlike C's int or Java's int).

# ── Integer literals in different bases ──
decimal = 255
binary  = 0b11111111    # Binary literal (prefix 0b)
octal   = 0o377         # Octal literal (prefix 0o)
hexa    = 0xFF          # Hexadecimal literal (prefix 0x)
print(f"decimal={decimal}, binary={binary}, octal={octal}, hex={hexa}")
# All four evaluate to 255

# ── Readability: underscores as digit separators (Python 3.6+) ──
population     = 8_000_000_000   # 8 billion — much clearer than 8000000000
light_speed    = 299_792_458     # metres per second
pi_digits      = 3_141_592_653

print(f"World population: {population:,}")   # comma-formatted output

# ── Arbitrary precision ──
import math
print(f"2^1000 has {len(str(2**1000))} digits")          # 302 digits
print(f"1000! has {len(str(math.factorial(1000)))} digits")  # 2568 digits

# ── int() constructor ──
print(int("42"))        # 42   — string to int
print(int(3.99))        # 3    — truncates (does NOT round)
print(int(True))        # 1    — bool to int
print(int("0xFF", 16))  # 255  — hex string to int
print(int("11", 2))     # 3    — binary string to int
````

````python
# ============================================================
# FLOAT TYPE — float
# ============================================================
# IEEE 754 double-precision: 64 bits (1 sign, 11 exponent, 52 mantissa)
# Range: approx ±1.8 × 10^308; precision: ~15-17 significant digits

# ── Float literals ──
f1 = 3.14159
f2 = -0.001
f3 = 1.5e6       # Scientific notation: 1.5 × 10^6 = 1500000.0
f4 = 2.5E-3      # 2.5 × 10^-3 = 0.0025
print(f"{f1}, {f2}, {f3}, {f4}")

# ── float() constructor ──
print(float("3.14"))    # 3.14
print(float("inf"))     # inf
print(float("-inf"))    # -inf
print(float("nan"))     # nan
print(float(True))      # 1.0
print(float(42))        # 42.0

# ── Precision limitation ──
import sys
print(f"Float max       : {sys.float_info.max:.3e}")   # ~1.798e+308
print(f"Float epsilon   : {sys.float_info.epsilon}")   # smallest diff from 1.0
print(f"Float digits    : {sys.float_info.dig}")       # ~15 decimal digits

# ── Implicit type promotion: int + float → float ──
result = 5 + 2.0    # int + float → float (Python promotes automatically)
print(f"5 + 2.0 = {result}, type = {type(result)}")   # float
````

````python
# ============================================================
# STRING TYPE — str
# ============================================================
# A str is an IMMUTABLE sequence of Unicode code points.
# Python 3 strings are Unicode by default (unlike Python 2).

# ── String literals using different delimiters ──
s1 = "double quotes"
s2 = 'single quotes'          # Interchangeable
s3 = """triple double"""      # Multi-line capable
s4 = '''triple single'''           # Also multi-line capable

# ── Strings containing quotes ──
q1 = "It's a beautiful day"   # Escape apostrophe, or:
q2 = 'He said "hello"'         # Mix quote styles

# ── Multi-line string ──
haiku = """An old silent pond
A frog jumps into the pond
Splash! Silence again"""
print(haiku)

# ── str() constructor ──
print(str(42))          # "42"
print(str(3.14))        # "3.14"
print(str(True))        # "True"
print(str(None))        # "None"

# ── String is immutable — cannot change characters in place ──
word = "hello"
try:
    word[0] = "H"   # TypeError: 'str' object does not support item assignment
except TypeError as e:
    print(f"\nImmutability confirmed: {e}")

# ── To "modify" a string, create a new one ──
modified = "H" + word[1:]
print(f"Original: {word!r}, Modified: {modified!r}")
````

````python
# ============================================================
# BOOL TYPE — bool (subclass of int!)
# ============================================================
# bool is a direct subclass of int. This means:
#   True  == 1  (and True is an int internally)
#   False == 0  (and False is an int internally)
# This has surprising arithmetic consequences.

print(f"type(True)         : {type(True)}")          # <class 'bool'>
print(f"isinstance(True,int): {isinstance(True,int)}")  # True!
print(f"True == 1          : {True == 1}")            # True
print(f"False == 0         : {False == 0}")           # True
print(f"True + True        : {True + True}")          # 2
print(f"True * 10          : {True * 10}")            # 10
print(f"False + 5          : {False + 5}")            # 5

# ── bool() constructor ──
print("\nbool() conversion:")
print(f"bool(0)    = {bool(0)}")       # False
print(f"bool(1)    = {bool(1)}")       # True
print(f"bool(-5)   = {bool(-5)}")      # True  (any non-zero int is truthy)
print(f"bool(0.0)  = {bool(0.0)}")     # False
print(f"bool('')   = {bool('')}")      # False (empty string)
print(f"bool('hi') = {bool('hi')}")    # True

# ── Practical: counting True values in a list ──
results = [True, False, True, True, False]
print(f"\nNumber of True values: {sum(results)}")   # 3  (True counts as 1)

# ── Comparison operators produce bool values ──
print(f"5 > 3 produces type: {type(5 > 3)}")   # <class 'bool'>
````

````python
# ============================================================
# NONETYPE — None
# ============================================================
# None is Python's null value. It is a SINGLETON — there is
# exactly one None object in any Python programme.
# None represents "no value", "missing", or "not yet assigned".

x = None
print(f"Value  : {x}")
print(f"Type   : {type(x)}")          # <class 'NoneType'>
print(f"bool() : {bool(None)}")       # False — None is falsy

# ── Always use 'is None' to check for None, NOT '== None' ──
if x is None:
    print("x has no value assigned")   # Correct pattern
# NOT: if x == None  ← works but is not idiomatic and can behave unexpectedly
#                     with objects that override __eq__

# ── Functions return None implicitly when no return statement ──
def greet(name):
    print(f"Hello, {name}!")
    # No return statement → returns None implicitly

result = greet("Alice")
print(f"\ngreet() returned: {result!r}")   # None

# ── None as a default parameter sentinel (common Python pattern) ──
def connect(host, port=None):
    if port is None:
        port = 8080   # Apply default only when not explicitly passed
    print(f"Connecting to {host}:{port}")

connect("localhost")        # port defaults to 8080
connect("example.com", 443) # explicit port used
````

## 🔷 Section 3 — Type Introspection

Python is a **dynamically typed** language, which means type information is
attached to **objects** at runtime rather than to **names** at compile time.
Python provides several built-in tools to inspect types during execution.

| Function | Purpose | Example |
|----------|---------|---------|
| `type(obj)` | Returns the type object of `obj` | `type(42)` → `<class 'int'>` |
| `id(obj)` | Returns the memory address (integer) | `id(x)` → `140234567890` |
| `isinstance(obj, cls)` | Checks if `obj` is an instance of `cls` | `isinstance(True, int)` → `True` |
| `sys.getsizeof(obj)` | Returns memory size in bytes | `sys.getsizeof(0)` → `28` |

### `type()` vs `isinstance()`

- `type(x) == int` checks the **exact** type — fails for subclasses
- `isinstance(x, int)` checks the type **and all parent classes** — preferred for type-checking

```python
isinstance(True, int)       # True  — bool is a subclass of int
type(True) == int           # False — exact match fails (it's bool, not int)
type(True) == bool          # True  — exact match
```

---

````python
# ============================================================
# TYPE INTROSPECTION — type(), id(), isinstance()
# ============================================================

values = [42, 3.14, "hello", True, None, [1,2], (3,4), {5:6}]

print(f"{'Value':<12} {'type()':<20} {'id() last 6 digits'}")
print("-" * 50)
for v in values:
    print(f"{repr(v):<12} {str(type(v)):<20} {id(v) % 1_000_000:>18}")

print()

# ── isinstance() with inheritance ──
print("isinstance() checks — note bool is a subclass of int:")
print(f"isinstance(True, bool) : {isinstance(True, bool)}")   # True
print(f"isinstance(True, int)  : {isinstance(True, int)}")    # True  (inheritance!)
print(f"type(True) == int      : {type(True) == int}")        # False (exact match)
print(f"type(True) == bool     : {type(True) == bool}")       # True

print()

# ── isinstance() with a tuple of types — very useful for input validation ──
def describe(val):
    if isinstance(val, bool):           # Check bool FIRST (before int!)
        return f"{val!r} is a bool"
    elif isinstance(val, (int, float)):
        return f"{val} is numeric"
    elif isinstance(val, str):
        return f"'{val}' is a string"
    else:
        return f"{val!r} is of type {type(val).__name__}"

for v in [True, 42, 3.14, "hi", None]:
    print(describe(v))
````

````python
# ============================================================
# MEMORY SIZES WITH sys.getsizeof()
# ============================================================
import sys

# ── Memory sizes of basic objects ──
objects = [0, 1, 100, 1000, True, False, None,
           "", "a", "hello", "hello world",
           0.0, 3.14, [], [1], [1,2,3]]

print(f"{'Object':<20} {'Type':<12} {'Bytes':>6}")
print("-" * 42)
for obj in objects:
    print(f"{repr(obj):<20} {type(obj).__name__:<12} {sys.getsizeof(obj):>6}")

print()
# ── Key observations: ──
# 1. Python ints: 28 bytes for small values (object overhead), grow with size
# 2. Python strings: 49 bytes overhead + 1 byte per ASCII character
# 3. Empty list: 56 bytes; each element adds 8 bytes (pointer size)
# 4. True/False: 28 bytes (same as small int)
# 5. None: only 16 bytes — very lightweight

# ── Integer memory growth with size ──
print("Integer memory growth:")
for bits in [0, 30, 62, 94, 126]:
    n = 2**bits if bits > 0 else 0
    print(f"  2^{bits:<3} → {sys.getsizeof(n)} bytes")
````

## 🔷 Section 4 — Type Conversion (Casting)

### Explicit Conversion

Python does **not** automatically convert between types in most situations.
You must explicitly call a conversion function:

| From \ To | `int` | `float` | `str` | `bool` |
|-----------|-------|---------|-------|--------|
| `int` | — | `float(n)` | `str(n)` | `bool(n)` |
| `float` | `int(f)` **truncates** | — | `str(f)` | `bool(f)` |
| `str` | `int(s)` | `float(s)` | — | `bool(s)` |
| `bool` | `int(b)` | `float(b)` | `str(b)` | — |

### Implicit Conversion (Type Coercion)

Python performs **limited** implicit conversion:

| Expression | Result Type | Reason |
|-----------|-------------|--------|
| `int + float` | `float` | Precision preservation |
| `bool + int` | `int` | bool is a subclass of int |
| `bool + float` | `float` | Chained coercion |

Python does **not** implicitly convert `str` to `int` or vice versa — `"5" + 5` raises `TypeError`.

---

````python
# ============================================================
# EXPLICIT TYPE CONVERSION (CASTING)
# ============================================================

# ── String to numeric ──
print("--- String → Numeric ---")
print(int("42"))         # 42
print(int("-17"))        # -17
print(float("3.14"))     # 3.14
print(int("0b1010", 2))  # 10  (binary string)
print(int("0xFF", 16))   # 255 (hex string)

# ── Numeric to string ──
print("\n--- Numeric → String ---")
print(repr(str(42)))     # '42'
print(repr(str(3.14)))   # '3.14'
print(repr(str(True)))   # 'True'

# ── int() TRUNCATES floats — does not round ──
print("\n--- float → int (truncation, not rounding) ---")
print(int(3.99))    # 3   ← truncated toward zero
print(int(-3.99))   # -3  ← truncated toward zero (not -4!)
print(int(3.1))     # 3
import math
print(f"round(3.99) = {round(3.99)}")  # 4  ← round() for rounding
print(f"math.floor(-3.99) = {math.floor(-3.99)}")  # -4  ← floor toward -inf
print(f"math.ceil(-3.99)  = {math.ceil(-3.99)}")   # -3  ← ceiling

# ── Illegal conversions raise exceptions ──
print("\n--- Illegal Conversions ---")
try: int("hello")
except ValueError as e: print(f"ValueError: {e}")

try: int(None)
except TypeError as e: print(f"TypeError: {e}")
````

### 4.2 Truthy and Falsy Values

In Python, every value has an inherent boolean interpretation.
This is used in `if` statements, `while` loops, and logical expressions.

#### Falsy Values (evaluate to `False` in a boolean context)

| Value | Type |
|-------|------|
| `0` | `int` |
| `0.0` | `float` |
| `0j` | `complex` |
| `""` (empty string) | `str` |
| `[]` (empty list) | `list` |
| `()` (empty tuple) | `tuple` |
| `{}` (empty dict) | `dict` |
| `set()` (empty set) | `set` |
| `None` | `NoneType` |
| `False` | `bool` |

**Everything else is truthy** — any non-zero number, non-empty collection,
non-empty string, or any object not in the list above.

---

````python
# ============================================================
# TRUTHY AND FALSY VALUES
# ============================================================

falsy_values = [0, 0.0, 0j, "", [], (), {}, set(), None, False]
print("Falsy values:")
for v in falsy_values:
    print(f"  bool({repr(v):<12}) = {bool(v)}")

print()
truthy_values = [1, -1, 0.001, " ", [0], (None,), {"a": 1}, {0}, True]
print("Truthy values:")
for v in truthy_values:
    print(f"  bool({repr(v):<14}) = {bool(v)}")

print()
# ── Practical use: check if a variable has a "meaningful" value ──
username = ""   # Simulating no input
if username:
    print(f"Welcome, {username}!")
else:
    print("Username is empty — please enter your name.")

# ── Idiomatic Python: 'if x' instead of 'if x != 0' ──
items = []
if not items:           # Equivalent to: if len(items) == 0
    print("The list is empty — nothing to process.")

# ── None vs False: both falsy, but semantically different ──
score = None   # "not yet set"
score = 0      # "set to zero" — also falsy but means something different!
# Always test explicitly: if score is None  vs  if not score
````

````python
# ============================================================
# EXERCISE 4 — Truthy/Falsy Classifier
# ============================================================
# Write a function that classifies a value's truthiness and
# explains WHY it is truthy or falsy.

def classify(val):
    if val is None:
        reason = "None is always falsy"
    elif isinstance(val, bool):
        reason = f"bool literal is {'True' if val else 'False'}"
    elif isinstance(val, (int, float)) and val == 0:
        reason = "zero numeric value"
    elif isinstance(val, str) and val == "":
        reason = "empty string"
    elif isinstance(val, (list, tuple, dict, set)) and len(val) == 0:
        reason = f"empty {type(val).__name__}"
    elif bool(val):
        reason = "non-zero / non-empty value"
    else:
        reason = "unknown falsy reason"

    label = "TRUTHY" if val else "FALSY"
    print(f"  {repr(val):<15} → {label:<7} ({reason})")

test_vals = [None, False, True, 0, -1, "", "  ", [], [0], {}, {"a":1}]
print("Truthy/Falsy Classification:")
for v in test_vals:
    classify(v)
````

### Challenge 2 — Type-Safe Calculator
Write a type-safe `safe_divide(a, b)` function that:
- Returns `float` result if both inputs are numeric
- Returns `None` if `b` is zero (with a descriptive print)
- Returns `None` with an explanation if inputs are non-numeric
- Also handles the case where inputs are numeric strings (converts them first)

````python
# ============================================================
# CHALLENGE 1 — Integer Dissection Without String Conversion
# ============================================================

n = 123456789
original = n

digits = []
temp = abs(n)
while temp > 0:         # (Day 10 preview: while loop)
    digits.append(temp % 10)
    temp //= 10
digits.reverse()        # (Day 6: list methods)

print(f"Number       : {original}")
print(f"Digits       : {digits}")
print(f"Digit count  : {len(digits)}")
print(f"Digit sum    : {sum(digits)}")

product = 1
for d in digits:
    product *= d
print(f"Digit product: {product}")
print(f"Largest digit : {max(digits)}")
print(f"Smallest digit: {min(digits)}")

# ── Alternative: number of digits using logarithm ──
import math
if n != 0:
    count_log = math.floor(math.log10(abs(n))) + 1
    print(f"\nDigit count via log10: {count_log}")
````

````python
# ============================================================
# CHALLENGE 2 — Type-Safe Calculator
# ============================================================

def safe_divide(a, b):
    # Step 1: try converting string inputs to numeric
    for val, name in [(a, 'a'), (b, 'b')]:
        if isinstance(val, str):
            try:
                if isinstance(a, str): a = float(a)
                if isinstance(b, str): b = float(b)
            except ValueError:
                print(f"Error: '{val}' (argument {name}) is not a valid number.")
                return None

    # Step 2: confirm both are numeric after conversion
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        print(f"Error: Both arguments must be numeric. Got {type(a).__name__} and {type(b).__name__}.")
        return None

    # Step 3: check for zero division
    if b == 0:
        print(f"Error: Division by zero — b cannot be 0.")
        return None

    return float(a) / float(b)

# Test cases
test_cases = [
    (10, 3),
    (10, 0),
    ("15.5", "2.5"),
    ("hello", 5),
    (10, "abc"),
        (True, False),     # bool inputs (subclass of int)
]
for a, b in test_cases:
    result = safe_divide(a, b)
    if result is not None:
        print(f"safe_divide({a!r}, {b!r}) = {result:.4f}")
````

## 📝 Summary — Day 02

| Concept | Key Point |
|---------|-----------|
| **Variable = reference** | Names point to objects; `y = x` does NOT copy the object |
| **Naming (PEP 8)** | `snake_case` for vars/functions; `UPPER_CASE` for constants; `PascalCase` for classes |
| **Multiple assignment** | `a, b = b, a` swaps without a temp variable |
| **Augmented ops** | `x += n` is shorthand for `x = x + n` |
| **int** | Unlimited precision; literals in decimal/binary/octal/hex |
| **float** | IEEE 754 double; approximate; `math.isclose()` for comparisons |
| **str** | Immutable Unicode sequence; single/double/triple quotes |
| **bool** | Subclass of `int`; `True==1`, `False==0`; check `bool` BEFORE `int` |
| **None** | Singleton null value; always test with `is None`, not `== None` |
| **type() vs isinstance()** | `isinstance()` respects inheritance — prefer it |
| **int() truncates** | `int(3.9)` → `3`; use `round()` or `math.floor()` as needed |
| **Falsy values** | `0`, `0.0`, `""`, `[]`, `{}`, `()`, `set()`, `None`, `False` |
