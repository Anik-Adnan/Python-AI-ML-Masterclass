## Section 1 - String Internals and Creation

A `str` is an immutable sequence of Unicode code points. Python 3 stores strings
internally as Unicode, meaning every string can contain any character from any language.

### Quoting Styles

| Style | Syntax | Use Case |
|-------|--------|----------|
| Single quote | 'hello' | Default for short strings |
| Double quote | "hello" | When string contains apostrophes |
| Triple single | '''multi-line''' | Multi-line content |
| Triple double | three double quotes | Docstrings |

### Key Properties
- Immutable: every "modification" creates a new string
- Hashable: usable as dictionary keys and set members
- Iterable: can be iterated character-by-character
- Sequence: supports indexing, slicing, `in`, `len()`, `min()`, `max()`
---

````python
# ============================================================
# STRING CREATION
# ============================================================

s1 = 'single quotes'
s2 = "double quotes"
s3 = """triple double -
spans multiple lines"""

print(s1, s2, sep=" | ")
print(s3)

# Choosing quote style
apostrophe = "It's a beautiful day"
with_quotes = 'He said "hello"'

print(apostrophe)
print(with_quotes)

# Raw strings: r"..." - backslashes are literal
raw = r"C:\Users\Alice\Documents"
normal = "C:\\Users\\Alice\\Documents"
print(f"\nraw   : {raw}")
print(f"normal: {normal}")
print(f"Equal : {raw == normal}")

# Unicode characters
pi_sym = "\u03C0"
heart = "\u2665"
print(f"\nUnicode: pi={pi_sym}, heart={heart}")
````

## Section 3 - Indexing, Slicing, and Operators

Strings support the same indexing and slicing syntax as lists, with one
critical difference: strings are immutable, so you cannot change a
character in place.

```
s = "Python"
s[0]       -> 'P'
s[-1]      -> 'n'
s[1:4]     -> 'yth'
s[::-1]    -> 'nohtyP'  (reversed)
s[0] = 'J' -> TypeError!
```
---

````python
# ============================================================
# INDEXING, SLICING, OPERATORS ON STRINGS
# ============================================================

s = "Hello, Python!"
print(f"String: {s!r}  (length={len(s)})")

print(f"\ns[0]   = {s[0]!r}")
print(f"s[-1]  = {s[-1]!r}")
print(f"s[7]   = {s[7]!r}")

print(f"\ns[:5]   = {s[:5]!r}")
print(f"s[7:]   = {s[7:]!r}")
print(f"s[::2]  = {s[::2]!r}")
print(f"s[::-1] = {s[::-1]!r}")

a = "Hello"
b = " World"
print(f"\na + b    = {a + b!r}")
print(f"a * 3    = {(a * 3)!r}")
print(f"'ello' in a = {'ello' in a}")

print(f"\n'apple' < 'banana': {'apple' < 'banana'}")
print(f"'Z' < 'a': {'Z' < 'a'}")

try:
    s[0] = 'J'
except TypeError as e:
    print(f"\nImmutability: {e}")

modified = 'J' + s[1:]
print(f"New string: {modified!r}")
````

## Section 4 - String Methods: Case, Strip, Search, Split, Join, Replace

| Method | Description |
|--------|-------------|
| upper(), lower() | Case conversion |
| title(), capitalize() | Title and sentence case |
| strip(), lstrip(), rstrip() | Whitespace/character removal |
| split(sep), join(iterable) | Convert between str and list |
| replace(old,new) | Substring substitution |
| find(), index() | Search (find returns -1, index raises) |
| count() | Count occurrences |
| startswith(), endswith() | Prefix/suffix tests |
---

````python
# ============================================================
# STRING METHODS - Case and Strip
# ============================================================

s = "  Hello, World!  "
print(f"Original: {s!r}")

print("\nCase methods:")
print(f"upper()     : {s.upper()!r}")
print(f"lower()     : {s.lower()!r}")
print(f"title()     : {s.title()!r}")
print(f"capitalize(): {'hello world'.capitalize()!r}")
print(f"swapcase()  : {'Hello World'.swapcase()!r}")

print("\nStrip methods:")
print(f"strip()     : {s.strip()!r}")
print(f"lstrip()    : {s.lstrip()!r}")
print(f"rstrip()    : {s.rstrip()!r}")
print(f"strip(custom): {'xyz!hello!xyz'.strip('xyz!')!r}")

raw_input_sim = "   alice@example.com   "
clean = raw_input_sim.strip().lower()
print(f"\nCleaned input: {clean!r}")
````

````python
# ============================================================
# STRING METHODS - Split, Join, Replace
# ============================================================

csv = "Alice,25,Berlin,Engineer"
parts = csv.split(",")
print(f"split(',')  : {parts}")

sentence = "  hello   world   python  "
words = sentence.split()
print(f"split()     : {words}")

limited = "a:b:c:d:e".split(":", 2)
print(f"split with maxsplit: {limited}")

multi = "Line1\nLine2\r\nLine3\nLine4"
print(f"splitlines(): {multi.splitlines()}")

words2 = ["Python", "is", "powerful"]
print(f"\nspace join : {' '.join(words2)!r}")
print(f"dash join  : {'-'.join(words2)!r}")

row = ["Alice", "25", "Berlin", "Engineer"]
print(f"CSV row    : {','.join(row)!r}")

text = "the cat sat on the mat with the cat"
print(f"\nOriginal       : {text!r}")
print(f"replace cat  : {text.replace('cat', 'dog')!r}")
print(f"replace 1st    : {text.replace('cat', 'dog', 1)!r}")
````

````python
# ============================================================
# STRING METHODS - Finding and Checking
# ============================================================

text = "Python is a versatile language. Python is popular."

print(f"find(Python)    : {text.find('Python')}")
print(f"find(Python,1)  : {text.find('Python', 1)}")
print(f"find(Java)      : {text.find('Java')}")

try:
    print(f"index(language) : {text.index('language')}")
    text.index("Java")
except ValueError as e:
    print(f"index(Java)     : ValueError - {e}")

print(f"count(Python)   : {text.count('Python')}")

url = "https://www.example.com/page"
print(f"\nstartswith(https)  : {url.startswith('https')}")
print(f"endswith(.com/page): {url.endswith('.com/page')}")

tests = ["Hello123", "12345", "hello", "   ", ""]
print(f"\n{'String':<15} {'isalnum':>8} {'isalpha':>8} {'isdigit':>8} {'isspace':>8}")
print("-" * 55)
for s in tests:
    print(f"{s!r:<15} {str(s.isalnum()):>8} {str(s.isalpha()):>8} {str(s.isdigit()):>8} {str(s.isspace()):>8}")
````

## Section 5 - String Formatting

| Method | Syntax | Recommended? |
|--------|--------|---------------|
| % operator | "Hello %s" % name | Legacy only |
| .format() | "Hello {}".format(name) | Cross-version |
| f-strings | f"Hello {name}" | Default choice |

### Format Specifiers

| Specifier | Meaning |
|-----------|---------|
| >10 / <10 / ^10 | Right/left/centre align width 10 |
| .2f | Float with 2 decimals |
| , | Thousands separator |
| e | Scientific notation |
| b, o, x | Binary, octal, hex |
---

````python
# ============================================================
# STRING FORMATTING - All Three Styles
# ============================================================

name  = "Alice"
age   = 20
score = 95.678

print("--- % formatting ---")
print("Name: %s" % name)
print("Age: %d, Score: %.1f" % (age, score))

print("\n--- .format() ---")
print("Name: {}".format(name))
print("{} is {} years old".format(name, age))
print("{name} scored {score:.1f}".format(name=name, score=score))

print("\n--- f-strings ---")
print(f"Name: {name}")
print(f"{name} is {age} years old")
print(f"Score: {score:.2f}")
print(f"Score percent: {score/100:.0%}")
````

````python
# ============================================================
# f-STRING FORMAT SPECIFIERS - Deep Dive
# ============================================================

pi   = 3.141592653589793
pop  = 8162534981
val  = 0.00123456

print("--- Number Formatting ---")
print(f"Default  : {pi}")
print(f"2 d.p.   : {pi:.2f}")
print(f"Sci.not. : {val:.3e}")
print(f"Percent  : {0.875:.1%}")
print(f"Thousands: {pop:,}")

n = 255
print(f"\nDecimal : {n:d}")
print(f"Binary  : {n:b}")
print(f"Octal   : {n:o}")
print(f"Hex     : {n:x}")

print("\n--- Alignment ---")
rows = [("Laptop", 999.99, 45), ("Mouse", 29.95, 200), ("Keyboard", 79.50, 89)]
for product, price, stock in rows:
    print(f"{product:<20} {price:>10.2f} {stock:>8}")

print(f"\n{'Left':<20}|")
print(f"{'Center':^20}|")
print(f"{'Right':>20}|")
````

````python
# EXERCISE 1 - Password Validator
def validate_password(pwd):
    errors = []
    if len(pwd) < 8:
        errors.append("At least 8 characters required")
    if not any(c.isupper() for c in pwd):
        errors.append("At least one uppercase letter required")
    if not any(c.islower() for c in pwd):
        errors.append("At least one lowercase letter required")
    if not any(c.isdigit() for c in pwd):
        errors.append("At least one digit required")
    special = "!@#$%^&*()"
    if not any(c in special for c in pwd):
        errors.append("At least one special character required")
    return errors

passwords = ["abc", "Password", "password1", "P@ssw0rd", "Short1!"]
for pwd in passwords:
    errs = validate_password(pwd)
    status = "VALID" if not errs else "INVALID"
    print(f"  {pwd!r:<20} {status}")
    for e in errs:
        print(f"      - {e}")
````

````python
# EXERCISE 2 - Word Frequency Counter
def word_frequency(text):
    words = text.lower().split()
    cleaned = [w.strip(".,!?;:\"'()[]") for w in words]
    freq = {}
    for word in cleaned:
        if word:
            freq[word] = freq.get(word, 0) + 1
    return dict(sorted(freq.items(), key=lambda x: x[1], reverse=True))

text = """To be or not to be that is the question
Whether tis nobler in the mind to suffer the slings"""
freq = word_frequency(text)
print("Top 10 words:")
for word, count in list(freq.items())[:10]:
    bar = "#" * count
    print(f"  {word:<12} {count:>3}  {bar}")
````

````python
# EXERCISE 3 - Caesar Cipher
def caesar_encrypt(text, shift):
    result = []
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            shifted = (ord(char) - base + shift) % 26 + base
            result.append(chr(shifted))
        else:
            result.append(char)
    return "".join(result)

def caesar_decrypt(text, shift):
    return caesar_encrypt(text, -shift)

plaintext = "Hello, World! Python is Amazing."
shift     = 13
encrypted = caesar_encrypt(plaintext, shift)
decrypted = caesar_decrypt(encrypted, shift)

print(f"Plaintext  : {plaintext}")
print(f"Encrypted  : {encrypted}")
print(f"Decrypted  : {decrypted}")
print(f"Match      : {decrypted == plaintext}")
````

````python
# EXERCISE 4 - String Statistics
def string_stats(text):
    words = text.split()
    return {
        "chars_total"   : len(text),
        "chars_no_space": len(text.replace(" ", "")),
        "word_count"    : len(words),
        "uppercase"     : sum(1 for c in text if c.isupper()),
        "lowercase"     : sum(1 for c in text if c.islower()),
        "digits"        : sum(1 for c in text if c.isdigit()),
        "avg_word_len"  : round(sum(len(w.strip(".,!?")) for w in words)/len(words), 2) if words else 0,
        "longest_word"  : max((w.strip(".,!?") for w in words), key=len) if words else "",
        "unique_words"  : len(set(w.lower().strip(".,!?") for w in words)),
    }

sample = "Python 3.10 was released in 2021. It introduced structural pattern matching!"
stats  = string_stats(sample)
print(f"Text: {sample!r}")
print()
for key, val in stats.items():
    print(f"  {key:<20}: {val}")
````

````python
# EXERCISE 5 - Email Parser
def parse_email(email):
    email = email.strip().lower()
    if "@" not in email:
        return None, None, None
    local, domain = email.split("@", 1)
    if "." not in domain:
        return local, domain, None
    parts = domain.rsplit(".", 1)
    return local, parts[0], parts[1]

emails = [
    "Alice.Johnson@University.EDU",
    "bob@gmail.com",
    "invalid-email",
    "user@domain.co.uk",
]
print(f"{'Email':<35} {'Username':<15} {'Domain':<15} {'TLD'}")
print("-" * 75)
for email in emails:
    user, domain, tld = parse_email(email)
    print(f"{email.strip():<35} {str(user):<15} {str(domain):<15} {str(tld)}")
````

````python
# CHALLENGE 1 - Run-Length Encoding (RLE)
def rle_encode(text):
    if not text:
        return ""
    result = []
    count = 1
    for i in range(1, len(text)):
        if text[i] == text[i-1]:
            count += 1
        else:
            result.append(f"{count}{text[i-1]}" if count > 1 else text[i-1])
            count = 1
    result.append(f"{count}{text[-1]}" if count > 1 else text[-1])
    return "".join(result)

def rle_decode(encoded):
    result = []
    i = 0
    while i < len(encoded):
        if encoded[i].isdigit():
            num_str = ""
            while i < len(encoded) and encoded[i].isdigit():
                num_str += encoded[i]
                i += 1
            result.append(encoded[i] * int(num_str))
            i += 1
        else:
            result.append(encoded[i])
            i += 1
    return "".join(result)

tests = ["AAABBBCCCA", "ABCDEFG", "AAAAAAAAAA", "AABBAABB"]
for t in tests:
    enc = rle_encode(t)
    dec = rle_decode(enc)
    ratio = len(enc)/len(t)
    print(f"  {t:<15} -> {enc:<15} -> {dec:<15} ratio={ratio:.2f}")
````

````python
# CHALLENGE 2 - Simple Template Engine
import re

def render_template(template, variables):
    result = template
    for key, value in variables.items():
        placeholder = "{{" + key + "}}"
        result = result.replace(placeholder, str(value))
    unfilled = re.findall(r"\{\{(\w+)\}\}", result)
    if unfilled:
        print(f"  Warning: unfilled placeholders: {unfilled}")
    return result

template = """Dear {{name}},

Your order #{{order_id}} has been confirmed.
Total: ${{total}}

Best regards,
{{company}}"""

variables = {
    "name": "Alice Johnson",
    "order_id": "ORD-2024-9847",
    "total": 99.98,
    "company": "TechBooks Ltd",
}
print(render_template(template, variables))
````
