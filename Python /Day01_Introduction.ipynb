## 🔷 Section 1 — What Is Python?

### 1.1 Definition

Python is a **high-level, interpreted, dynamically typed, garbage-collected,
multi-paradigm programming language** created by **Guido van Rossum** and first
released in **1991**. Its design philosophy centres on code readability and
expressiveness, encapsulated by the principle:

> *"There should be one — and preferably only one — obvious way to do it."*

### 1.2 Core Language Characteristics

| Property | Description |
|----------|-------------|
| **High-level** | Abstracts away hardware details (memory management, CPU registers) |
| **Interpreted** | Executed line-by-line by the CPython interpreter at runtime — no separate compile step |
| **Dynamically typed** | Variable types are inferred at runtime; no explicit type declarations required |
| **Garbage-collected** | Memory is automatically reclaimed when objects are no longer referenced |
| **Multi-paradigm** | Supports procedural, object-oriented, and functional programming styles |
| **Cross-platform** | Identical source code runs on Windows, macOS, Linux, and ARM |
| **Batteries included** | Extensive standard library ships with every Python installation |

### 1.3 Python 2 vs Python 3

This course uses **Python 3.10+**. Python 2 reached **end-of-life on January 1, 2020**.

| Feature | Python 2 | Python 3 |
|---------|----------|----------|
| `print` | Statement: `print "hi"` | Function: `print("hi")` |
| Integer division | `5/2 → 2` (truncates) | `5/2 → 2.5` (exact) |
| Unicode strings | External — `u"hello"` | Native default |
| `range()` | Returns a full list in memory | Returns a lazy iterator |

### 3.1 Indentation — Python's Most Important Syntax Rule

In C, Java, and JavaScript, **curly braces `{}`** delimit code blocks. Python
instead uses **whitespace indentation**, and this is **enforced by the interpreter**,
not just a style recommendation.

Incorrect indentation raises an `IndentationError` that prevents your programme from running.

#### PEP 8 Standard for Indentation

**PEP 8** (Python Enhancement Proposal 8) is Python's official style guide. It mandates:
- Exactly **4 spaces** per indentation level
- **Never mix tabs and spaces** (Python 3 raises a `TabError` if you do)
- Use your editor's "convert tabs to spaces" setting

#### Why Did Guido Make Indentation Mandatory?

Guido van Rossum deliberately enforced indentation because studies of programmer
behaviour showed that developers *always* indent their code to reflect logical
structure. Making it mandatory eliminates an entire class of bugs where code
indentation suggests one meaning but mismatched braces impose another.

```
# Correct — body is indented 4 spaces:
if condition:
    statement_1    ← inside the block
    statement_2    ← inside the block
statement_3        ← outside the block (back to level 0)

# Wrong — will raise IndentationError:
if condition:
statement_1        ← ERROR: expected an indented block
```

---

````python
# ============================================================
# INDENTATION IN PRACTICE
# ============================================================

x = 10   # A variable assignment — will be covered in detail on Day 2

# ── Single-level indentation ──
if x > 5:                             # Header line ends with ':'
    print("x is greater than 5")     # 4-space indent → inside the if-block
    print("This also runs when x > 5")  # Same indent level → same block

print("This always runs")            # No indent → outside all blocks

# ── Nested indentation (two levels) ──
if x > 5:                            # Level 0 header
    print("x > 5")                   # Level 1 (4 spaces)
    if x > 8:                        # Level 1 — nested header
        print("x > 8 too")           # Level 2 (8 spaces)

# ── Visual map of block structure ──
# if x > 5:             ← column 0
#     [BLOCK A]         ← column 4
#     if x > 8:         ← column 4 (still in outer block)
#         [BLOCK B]     ← column 8 (inner nested block)
# [outside]             ← column 0 (back to global scope)

# ── Demonstrating an IndentationError safely ──
# The following is commented out — remove # to observe the error:
# if True:
# print("IndentationError here")   ← would raise IndentationError
````

````python
# ============================================================
# CASE SENSITIVITY IN PYTHON
# ============================================================
# Every character of an identifier is significant, including its case.
# Python distinguishes 'a' from 'A' in variable names, function names,
# module names, class names, and keywords.

# These are FOUR completely independent variables:
name  = "Alice"    # lowercase — standard variable
Name  = "Bob"      # capitalised first letter
NAME  = "CHARLIE"  # all-uppercase (used by convention for CONSTANTS)
nAmE  = "Dave"     # mixed case — strongly discouraged by PEP 8

print(name)    # Alice
print(Name)    # Bob
print(NAME)    # CHARLIE
print(nAmE)    # Dave

# ── Built-in names are also case-sensitive ──
print("hello")   # ✓ Works
# Print("hello")   # ✗ NameError: name 'Print' is not defined
# PRINT("hello")   # ✗ NameError: name 'PRINT' is not defined

# ── PEP 8 Naming Conventions ──────────────────────────────
# These are not enforced by Python, but are universally followed in industry:
#
#   Variables & functions : lowercase_with_underscores    (snake_case)
#   Constants             : ALL_CAPS_WITH_UNDERSCORES
#   Classes               : CapitalisedWords              (PascalCase / UpperCamelCase)
#   Private names         : _leading_underscore
#   "Dunder" methods      : __double_underscores__

student_name = "Alice"       # ✓ snake_case
MAX_STUDENTS = 30            # ✓ constant in ALL_CAPS
PI           = 3.14159265    # ✓ mathematical constant
````

### 3.2 Statements, Expressions, and Line Continuation

#### Statements vs Expressions

| Term | Definition | Example |
|------|-----------|---------|
| **Expression** | A code fragment that **evaluates to a value** | `3 + 4`, `"hello"`, `x * 2` |
| **Statement** | A complete instruction that **performs an action** | `x = 5`, `print("hi")` |

In Python, one statement occupies one line. There are two ways to break a long
statement across multiple lines:

#### Explicit Continuation — Backslash `\`
```
total = 100 + 200 + 300 + \
        400 + 500 + 600
```
The backslash tells Python "this line continues on the next."

#### Implicit Continuation — Inside Brackets (Preferred)
```
total = (100 + 200 + 300 +
         400 + 500 + 600)
```
Inside any `( )`, `[ ]`, or `{ }`, Python automatically treats the expression
as continuing until the closing bracket. This is the **preferred** PEP 8 approach.

#### Multiple Statements on One Line (Discouraged)
```python
a = 1; b = 2; c = 3   # Legal but hurts readability — avoid this
```

---

````python
# ============================================================
# STATEMENTS, EXPRESSIONS, AND LINE CONTINUATION
# ============================================================

# ── Expressions vs statements ──
3 + 4           # Expression: evaluates to 7 (Jupyter displays it; scripts ignore it)
x = 3 + 4       # Statement: assigns the value 7 to variable x
print(x)        # Statement: calls print() as a side-effect (output to screen)

# ── Explicit line continuation using backslash ──
total_explicit = 100 + 200 + 300 +                  400 + 500 + 600
print("Explicit continuation:", total_explicit)   # 2100

# ── Implicit continuation inside parentheses (preferred) ──
total_implicit = (100 + 200 + 300 +
                  400 + 500 + 600)
print("Implicit continuation:", total_implicit)   # 2100

# ── Implicit continuation inside square brackets (lists — Day 6) ──
scores = [95, 87, 92,
          88, 76, 100]    # One list split across two lines
print("Scores:", scores)

# ── Multiple statements on one line — legal but discouraged ──
a = 1; b = 2; c = 3
print(a, b, c)    # 1 2 3

# ── Preferred: one statement per line ──
a = 1
b = 2
c = 3
print(a, b, c)    # Same result — but more readable
````

## 🔷 Section 4 — Comments

### What Is a Comment?

A **comment** is text in source code that is **completely ignored by the Python
interpreter**. Comments exist for the human reader only — to explain intent,
document logic, or temporarily disable code during debugging.

### Types of Comments in Python

| Type | Syntax | Primary Use |
|------|--------|------------|
| **Single-line** | `# text` | Explain a specific line or statement |
| **Multi-line** | Multiple `#` lines | Explain a block of logic |
| **Docstring** | `"""text"""` (triple-quoted) | Document a module, class, or function |

### The Golden Rule: Comment the WHY, Not the WHAT

```python
# ✗ BAD — states what is already obvious from the code:
x = x + 1    # Add 1 to x

# ✓ GOOD — explains the reasoning behind the code:
x = x + 1    # Offset by 1 to convert from 0-based array index to 1-based display position
```

**Principles for good comments:**
- Comment every non-trivial block of logic
- Keep comments synchronised with the code — stale comments are more dangerous than no comments
- Use complete sentences beginning with a capital letter
- Prefer `#` for inline and block comments; reserve docstrings for function/class/module documentation

---

````python
# ============================================================
# COMMENTS IN PYTHON — Practical Examples
# ============================================================

# This is a single-line comment.
# Everything after '#' on a line is ignored by the interpreter.

x = 42   # Inline comment: appears after code on the same line

# ── Multi-line comment using consecutive single-line comments ──
# The following block computes the area of a rectangle.
# Both dimensions are assumed to be positive integers.
# Formula: area = width * height
width  = 8
height = 5
area   = width * height
print("Area:", area)    # Output: Area: 40

# ── Triple-quoted strings as multi-line content ──
# When placed at the start of a module, class, or function,
# triple-quoted strings become "docstrings" (accessed via help()).
# For now, observe that they can span multiple lines:

description = """
This is a multi-line string.
It preserves line breaks exactly as written.
Python stores it as a single string object with embedded newlines.
"""
print(description)

# ── Temporarily disabling code with comments ("commenting out") ──
result = 10 * 5
# result = 10 / 0   ← Commented out: would raise ZeroDivisionError
print("Result:", result)    # Output: Result: 50

# ── Note on docstrings (preview — covered in depth on Day 11) ──
# def greet(name):
#     """Return a personalised greeting for the given name."""
#     return f"Hello, {name}!"
# The triple-quoted string above is the function's docstring.
````

## 🔷 Section 5 — The `print()` Function

### Complete Function Signature

```python
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
```

| Parameter | Default Value | Description |
|-----------|--------------|-------------|
| `*objects` | *(required)* | One or more comma-separated values to output |
| `sep` | `' '` (one space) | String inserted **between** consecutive objects |
| `end` | `'\n'` (newline) | String appended **after** the last object |
| `file` | `sys.stdout` | Output destination (terminal, file, etc.) |
| `flush` | `False` | If `True`, forcibly flush the output buffer immediately |

### print() vs return — A Critical Distinction

`print()` sends a value to the screen but the value cannot be used in further computation.
`return` (covered on Day 11) sends a value back to the caller for further processing.
Confusing the two is among the most common conceptual errors for beginners.

```python
# print() — side-effect only; result cannot be captured:
print(2 + 3)     # Displays 5, but you cannot do: x = print(2 + 3) meaningfully
                  # (x would be None, the implicit return value of print)

# return — value available for further computation:
def add(a, b):
    return a + b
result = add(2, 3)   # result = 5 — can be used in expressions
```

---

````python
# ============================================================
# THE print() FUNCTION — Basic Usage
# ============================================================

# ── Printing different data types ──
print("Hello, World!")   # str  (string)
print(42)                # int  (integer)
print(3.14159)           # float
print(True)              # bool
print(None)              # NoneType — represents "absence of a value"

print()   # ← Prints an empty line (just the default newline)

# ── Printing multiple values in one call ──
# Default separator between arguments is a single space
print("Name:", "Alice", "Age:", 21)   # Name: Alice Age: 21

# ── Printing variables ──
student = "Bob"
grade   = 95
print("Student:", student, "| Grade:", grade)

# ── Printing computed expressions directly ──
print("Sum of 7 and 13:", 7 + 13)   # Sum of 7 and 13: 20

# ── String concatenation inside print (requires all parts to be str) ──
first = "Python"
second = "is elegant"
print(first + " " + second)   # Python is elegant
# Caution: print("Age: " + 21) raises TypeError — int cannot concat with str.
# Use f-strings or the comma syntax instead (shown in the next cells).
````

````python
# ============================================================
# print() — sep AND end PARAMETERS
# ============================================================

# ── sep: string inserted between each argument ──
print("Python", "Java", "C++",      sep=", ")    # Python, Java, C++
print("2024",   "06",   "15",       sep="-")     # 2024-06-15
print("A",      "B",    "C",        sep=" → ")   # A → B → C

print()   # blank line

# ── sep="\n" prints each argument on its own line ──
print("First line", "Second line", "Third line", sep="\n")
# Output:
# First line
# Second line
# Third line

print()

# ── end: string appended after the last argument (replaces default newline) ──
# Default end="\n" means print() always moves to the next line.
# Setting end="" suppresses the automatic newline.
print("Loading",    end="")      # No newline — cursor stays on same line
print(".",          end="")
print(".",          end="")
print(".",          end="")
print(" Complete!")              # end="\n" (default) — newline here
# Output: Loading... Complete!

print()

# ── Combining sep and end ──
print("Step 1", "Step 2", "Step 3", sep=" → ", end=" ✓\n")
# Output: Step 1 → Step 2 → Step 3 ✓

# ── Practical: right-aligned table using end ──
print("Item",  end="\t")   # Escape sequence \t = tab character
print("Price", end="\t")
print("Stock")
print("-" * 30)
print("Laptop",  end="\t"); print("$999", end="\t"); print(45)
print("Mouse",   end="\t"); print("$29",  end="\t"); print(200)
````

````python
# ============================================================
# print() — F-STRINGS (Formatted String Literals)
# ============================================================
# Introduced in Python 3.6 (PEP 498), f-strings are the MODERN
# and PREFERRED way to embed expressions inside string literals.
#
# Syntax: prefix the string with 'f' (or 'F') and use {expr}
# anywhere in the string. The expression is evaluated at runtime
# and its string representation is substituted in place.

name  = "Alice"
age   = 20
gpa   = 3.875

# ── Basic substitution ──
print(f"Name: {name}")                     # Name: Alice
print(f"Age: {age}")                       # Age: 20
print(f"{name} is {age} years old.")       # Alice is 20 years old.

# ── Arbitrary expressions inside braces ──
print(f"Next year age: {age + 1}")         # Next year age: 21
print(f"5 squared: {5 ** 2}")             # 5 squared: 25
print(f"Name in uppercase: {name.upper()}")  # Name in uppercase: ALICE

# ── Format specifiers (after a colon inside the braces) ──
pi = 3.141592653589793
print(f"pi to 2 d.p.  : {pi:.2f}")    # 3.14
print(f"pi to 5 d.p.  : {pi:.5f}")    # 3.14159
print(f"GPA           : {gpa:.2f}")   # 3.88

# ── Width and alignment ──
# < left-align   > right-align   ^ centre-align   (followed by field width)
print(f"{'Product':<15} {'Price':>8} {'Stock':>6}")
print(f"{'Laptop':<15} {'$999.99':>8} {45:>6}")
print(f"{'Wireless Mouse':<15} {'$29.95':>8} {200:>6}")

# ── Note: f-strings will be explored further on Day 5 (Strings) ──
````

## 🔷 Section 6 — Expressions, Operators, and Arithmetic

### What Is an Expression?

An **expression** is any fragment of code that **evaluates to a value**:

```
42              → the integer 42
3.14            → the float 3.14
"hello"         → the string 'hello'
2 + 3           → evaluates to 5
x * y - 1      → depends on x and y at runtime
```

### Python's Arithmetic Operators

| Operator | Name | Example | Result | Notes |
|----------|------|---------|--------|-------|
| `+` | Addition | `7 + 3` | `10` | |
| `-` | Subtraction | `7 - 3` | `4` | |
| `*` | Multiplication | `7 * 3` | `21` | |
| `/` | True Division | `7 / 2` | `3.5` | **Always returns `float`** |
| `//` | Floor Division | `7 // 2` | `3` | Truncates toward −∞ |
| `%` | Modulo | `7 % 2` | `1` | Remainder after floor division |
| `**` | Exponentiation | `2 ** 10` | `1024` | Right-associative |

### Operator Precedence (PEMDAS / BODMAS)

From highest to lowest:

1. `**` — Exponentiation *(right-to-left)*
2. `+x`, `-x` — Unary plus / minus
3. `*`, `/`, `//`, `%` — Multiplicative *(left-to-right)*
4. `+`, `-` — Additive *(left-to-right)*

When in doubt, use parentheses — they improve readability *and* guarantee correctness.

````python
# ============================================================
# ARITHMETIC EXPRESSIONS — Comprehensive Examples
# ============================================================
import math   # Standard library module (covered in depth on Day 18)

# ── All seven arithmetic operators ──
print("Addition       :", 10 + 3)    # 13
print("Subtraction    :", 10 - 3)    # 7
print("Multiplication :", 10 * 3)    # 30
print("True Division  :", 10 / 3)    # 3.3333333333333335  (always a float!)
print("Floor Division :", 10 // 3)   # 3   (integer quotient, rounds toward -inf)
print("Modulo         :", 10 % 3)    # 1   (remainder)
print("Exponentiation :", 2 ** 10)   # 1024

print()

# ── Type produced by division operators ──
print(type(10 / 2))    # <class 'float'>  — even when divisible exactly
print(type(10 // 2))   # <class 'int'>    — floor division on ints → int

print()

# ── Practical: check even / odd with modulo ──
n = 17
parity = "even" if n % 2 == 0 else "odd"  # Ternary expression (preview of Day 9)
print(f"{n} is {parity}")    # 17 is odd

# ── Practical: extract digits ──
number = 4567
units    = number % 10           # 7
tens     = (number // 10)  % 10  # 6
hundreds = (number // 100) % 10  # 5
thousands = number // 1000       # 4
print(f"Digits of {number}: {thousands}-{hundreds}-{tens}-{units}")

# ── Python integers have UNLIMITED precision ──
print(f"2^64  = {2**64}")
print(f"2^128 = {2**128}")
print(f"20!   = {math.factorial(20)}")   # 2432902008176640000
print(f"100! has {len(str(math.factorial(100)))} digits")  # 158 digits
````

````python
# ============================================================
# OPERATOR PRECEDENCE — PEMDAS / BODMAS
# ============================================================

# ── Multiplication before addition ──
r1 = 2 + 3 * 4        # 3*4=12, then 2+12=14
r2 = (2 + 3) * 4      # 2+3=5, then 5*4=20
print(f"2 + 3 * 4   = {r1}")    # 14
print(f"(2 + 3) * 4 = {r2}")    # 20

# ── Exponentiation before multiplication ──
r3 = 2 * 3 ** 2       # 3**2=9, then 2*9=18
r4 = (2 * 3) ** 2     # 2*3=6, then 6**2=36
print(f"2 * 3 ** 2   = {r3}")   # 18
print(f"(2 * 3) ** 2 = {r4}")   # 36

# ── Exponentiation is RIGHT-associative ──
r5 = 2 ** 3 ** 2       # = 2 ** (3**2) = 2**9 = 512
r6 = (2 ** 3) ** 2     # = 8**2 = 64
print(f"2 ** 3 ** 2    = {r5}")  # 512  (right-to-left)
print(f"(2 ** 3) ** 2  = {r6}")  # 64

# ── Complex mixed expression — trace step by step ──
expr = 10 + 2 * 6 - 4 / 2 + 3 ** 2
# Evaluation order:
#  1. 3 ** 2    = 9
#  2. 2 * 6     = 12
#  3. 4 / 2     = 2.0
#  4. 10 + 12   = 22
#  5. 22 - 2.0  = 20.0
#  6. 20.0 + 9  = 29.0
print(f"10 + 2*6 - 4/2 + 3**2 = {expr}")   # 29.0

# ── Best practice: use parentheses for clarity ──
clear = ((10 + (2 * 6)) - (4 / 2)) + (3 ** 2)
print(f"Same with explicit parentheses = {clear}")   # 29.0
````

````python
# ============================================================
# FLOOR DIVISION ( // ) AND MODULO ( % ) — DEEP DIVE
# ============================================================

# ── Floor division rounds toward NEGATIVE INFINITY (not toward zero) ──
print("Positive values:")
print(17 // 5)      #  3  (5 fits into 17 three times)
print(-17 // 5)     # -4  (NOT -3! floored toward -inf)
print(17 // -5)     # -4  (same rule applies when divisor is negative)

print("\nModulo — remainder sign matches the divisor:")
print(17 % 5)       #  2  (17 = 3*5 + 2)
print(-17 % 5)      #  3  (NOT -2! Python modulo matches sign of divisor)
print(17 % -5)      # -3

# ── Invariant: a == (a // b) * b + (a % b) — always true ──
a, b = 17, 5
verify = (a // b) * b + (a % b)
print(f"\nInvariant check: {a} == (17//5)*5 + 17%5 = {verify} → {a == verify}")

# ── Application 1: Time Conversion (seconds → h:m:s) ──
total_seconds = 7384
hours   = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60
print(f"\n{total_seconds}s = {hours}h {minutes}m {seconds}s")  # 2h 3m 4s

# ── Application 2: Check leap year divisibility rules ──
year = 2024
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"\n{year} is a leap year")   # 2024 is a leap year
else:
    print(f"\n{year} is not a leap year")

# ── Application 3: Cyclic indexing (day-of-week calculation) ──
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
start_day_index = 0    # Monday = index 0
days_later      = 17   # What day is 17 days after Monday?
target = days[(start_day_index + days_later) % 7]
print(f"17 days after Monday is: {target}")   # Thursday
````

````python
# ============================================================
# FLOATING-POINT ARITHMETIC AND PRECISION
# ============================================================
# Floats use the IEEE 754 double-precision (64-bit) standard.
# 64 bits = 1 sign + 11 exponent + 52 mantissa bits.
# Most decimal fractions cannot be represented exactly in binary.

# ── The famous floating-point surprise ──
print(0.1 + 0.2)            # 0.30000000000000004  (not 0.3!)
print(0.1 + 0.2 == 0.3)    # False — a fundamental IEEE 754 artefact

# This is NOT a Python bug. It affects C, Java, JavaScript, and all
# languages using IEEE 754 floating-point arithmetic.

import math

# ── Workaround 1: round() for display purposes ──
result = 0.1 + 0.2
print(round(result, 10))          # 0.3  (rounded to 10 decimal places)
print(round(result, 10) == 0.3)   # True

# ── Workaround 2: math.isclose() for comparisons (recommended) ──
# isclose(a, b, rel_tol=1e-9, abs_tol=0.0)
print(math.isclose(0.1 + 0.2, 0.3))              # True
print(math.isclose(0.1 + 0.2, 0.3, abs_tol=1e-9))  # True

# ── Workaround 3: decimal.Decimal for exact decimal arithmetic ──
from decimal import Decimal, getcontext
getcontext().prec = 28   # Set precision to 28 significant digits
print(Decimal('0.1') + Decimal('0.2'))   # 0.3  (exact!)

# ── Special float values ──
print("\nSpecial float values:")
print(math.inf)            # inf   — positive infinity
print(-math.inf)           # -inf  — negative infinity
print(math.nan)            # nan   — Not a Number (undefined result)
print(math.isnan(math.nan))   # True
print(math.isinf(math.inf))   # True
````

## 🔷 Section 7 — The `input()` Function

### Reading User Input at Runtime

`input()` **suspends programme execution** and waits for the user to type text
and press Enter. It then returns the entered text as a **string**.

### Function Signature

```python
input(prompt='')
```

| Parameter | Description |
|-----------|-------------|
| `prompt` | Optional string displayed to the user before input (default: `''`) |
| **Return type** | Always `str` — regardless of what the user types |

### ⚠️ Critical Rule: Always Convert the Result

Because `input()` invariably returns a string, **you must convert** the value
when your programme requires a number:

```python
age = input("Enter your age: ")    # Returns "20" — a STRING, not an integer
age = int(age)                     # Explicit conversion: "20" → 20

# Shorthand (single line):
age = int(input("Enter your age: "))
```

### Type Conversion Functions

| Target Type | Function | Example | Result |
|-------------|----------|---------|--------|
| Integer | `int()` | `int("42")` | `42` |
| Float | `float()` | `float("3.14")` | `3.14` |
| String | `str()` | `str(42)` | `"42"` |
| Boolean | `bool()` | `bool(0)` | `False` |

## 🔷 Section 8 — Common Beginner Errors

Reading error messages carefully is one of the most valuable skills you can develop.
Python's error output is deliberately informative — it tells you **exactly where** and
**why** something failed.

### The Six Most Common Error Types (Day 1 Level)

| Error Type | When It Occurs | Minimal Example |
|------------|---------------|-----------------|
| `SyntaxError` | Code violates Python grammar | `print("hello"` — missing `)` |
| `IndentationError` | Block not indented correctly | Missing indent after `if x:` |
| `NameError` | Variable used before assignment | `print(x)` when `x` never defined |
| `TypeError` | Wrong type for an operation | `"age: " + 25` — str + int |
| `ValueError` | Correct type, wrong value | `int("hello")` |
| `ZeroDivisionError` | Division by zero | `10 / 0` |

### Anatomy of a Python Traceback

```
Traceback (most recent call last):
  File "example.py", line 5, in <module>
    result = 10 / 0
ZeroDivisionError: division by zero
      ↑                ↑
  Error type      Human-readable description
```

**Read tracebacks from bottom to top**: the last line is the actual error.
The lines above show the call chain that led to the error.

by repeatedly summing the digits until a single digit remains.

````python
# ============================================================
# CHALLENGE 1 — Fibonacci Sequence (First 15 Terms, No Loops)
# ============================================================
# Each term = sum of the two preceding terms.
# We manually assign each variable — this is deliberately tedious
# to illustrate exactly WHY iteration (Day 10) is indispensable.

f0  = 0
f1  = 1
f2  = f0  + f1   # 1
f3  = f1  + f2   # 2
f4  = f2  + f3   # 3
f5  = f3  + f4   # 5
f6  = f4  + f5   # 8
f7  = f5  + f6   # 13
f8  = f6  + f7   # 21
f9  = f7  + f8   # 34
f10 = f8  + f9   # 55
f11 = f9  + f10  # 89
f12 = f10 + f11  # 144
f13 = f11 + f12  # 233
f14 = f12 + f13  # 377

print("First 15 Fibonacci Numbers (F(0) through F(14)):")
print(f"F(0)  = {f0}")
print(f"F(1)  = {f1}")
print(f"F(2)  = {f2}")
print(f"F(3)  = {f3}")
print(f"F(4)  = {f4}")
print(f"F(5)  = {f5}")
print(f"F(6)  = {f6}")
print(f"F(7)  = {f7}")
print(f"F(8)  = {f8}")
print(f"F(9)  = {f9}")
print(f"F(10) = {f10}")
print(f"F(11) = {f11}")
print(f"F(12) = {f12}")
print(f"F(13) = {f13}")
print(f"F(14) = {f14}")

# ── Verify: F(15) = F(13) + F(14) = 377 + 233 = 610 ──
print(f"\nF(15) = F(13) + F(14) = {f13} + {f14} = {f13 + f14}")

# ── Reflection ──
# Notice how tedious this is! On Day 10 you will write the same
# computation in 3 lines using a while-loop. That insight is the
# central motivation for learning iteration.
````

````python
# ============================================================
# CHALLENGE 2 — Digital Root Calculator
# ============================================================
# Digital root: sum digits repeatedly until a single digit remains.
# Direct formula (for n > 0): digital_root(n) = 1 + (n - 1) % 9

n = 9875

print(f"Analysing n = {n}")
print("=" * 40)

# ── Extract all four digits using floor division and modulo ──
units     = n % 10              # 5
tens      = (n //   10) % 10   # 7
hundreds  = (n //  100) % 10   # 8
thousands = (n // 1000) % 10   # 9
print(f"Digits: {thousands}, {hundreds}, {tens}, {units}")

# ── Sum of digits ──
digit_sum = thousands + hundreds + tens + units
print(f"Sum of digits: {thousands}+{hundreds}+{tens}+{units} = {digit_sum}")  # 29

# ── Second-level sum (29 → 2+9 = 11) ──
sum2_units = digit_sum % 10
sum2_tens  = digit_sum // 10
level2 = sum2_tens + sum2_units
print(f"Level 2 sum  : {sum2_tens}+{sum2_units} = {level2}")    # 11 → 1+1 = 2

# ── Third-level sum (11 → 1+1 = 2) ──
level3_units = level2 % 10
level3_tens  = level2 // 10
level3 = level3_tens + level3_units
print(f"Level 3 sum  : {level3_tens}+{level3_units} = {level3}")   # 2

# ── Direct formula verification ──
digital_root_formula = 1 + (n - 1) % 9
print(f"\nFormula result: 1 + ({n}-1) % 9 = {digital_root_formula}")
print(f"Iterative result: {level3}")
print(f"Match: {digital_root_formula == level3}")   # True

# ── Why does the formula work? ──
# It is based on the 'casting out nines' property of decimal numbers:
# any integer is congruent to the sum of its digits modulo 9.
print(f"\nVerification: {n} % 9 = {n % 9}")   # Same as digital_root % 9
````
